## 1. Extensión de archivos y Punto de Entrada

Los scripts en Python se guardan con la extensión **`.py`**. Son archivos de texto que son interpretados por el intérprete de Python (Corcuera, IntroPython).

Cuando ejecutas un archivo directamente (por ejemplo: `python script.py`), Python asigna un nombre especial a ese módulo: `"__main__"`.

Esto nos permite definir un **punto de entrada** explícito, similar a la función `main()` en C.

> **Puente con C:** En C, el punto de entrada es siempre la función `main()`. En Python, el código se ejecuta de arriba a abajo, pero la convención `if __name__ == '__main__':` nos permite controlar qué código se ejecuta solo cuando el archivo es el programa principal (y no cuando es importado como módulo).

```python
# mi_modulo.py
def main():
    print("¡Este es el punto de entrada principal! Solo se ejecutará si este archivo se ejecuta directamente.")

print("¡Este código se ejecutará incluso si se importa como módulo!")

if __name__ == '__main__':
    main()
```

## 2. La función `print()`

Antes de sumergirnos en tipos y operadores, conozcamos la herramienta que usaremos para ver resultados: la función `print()`.

`print()` es una **función predefinida** (built-in) que envía texto a la salida estándar (la consola). Por defecto, a diferencia de C, python importa automáticamente la librería estándar. Sus características más comunes son:

| Uso | Ejemplo | Resultado |
|-----|---------|-----------|
| **Un solo valor** (agrega el salto de línea por defecto) | `print("Hola")` | `Hola` |
| **Varios valores** (separados por espacio) | `print("Hola", "Mundo")` | `Hola Mundo` |
| **Parámetro `sep`** (cambia el separador) | `print("a", "b", "c", sep="-")` | `a-b-c` |
| **Parámetro `end`** (cambia el final, por defecto `\n`) | `print("Hola", end=" ")` + `print("Mundo")` | `Hola Mundo` |
| **Formato con f-strings** | `print(f"2 + 2 = {2 + 2}")` | `2 + 2 = 4` |
| **Representación con `!r`** | `print(f"{'Hola'!r}")` | `'Hola'` |

> **Nota:** En una clase futura sobre funciones exploraremos `print()` a fondo y veremos todas las funciones predefinidas. Por ahora, con estos usos básicos tenemos suficiente para lo que viene.

Ejecuta la celda de abajo para ver `print()` en acción:

```python
print("Hola")                          # Un solo valor
print("Hola", "Mundo")               # Varios valores
print("a", "b", "c", sep="-")      # Separador personalizado
print("Hola", end=" ")               # Sin salto de línea al final
print("Mundo")                        # Continúa en la misma línea
print(f"2 + 2 = {2 + 2}")            # f-string
print(f"{'Python'!r}")               # Representación con comillas
```

## 3. Tipado dinámico y Modelo de Objetos

### Todo es un Objeto

En Python, **todo es un objeto** (incluso las funciones y los números). Cada objeto se caracteriza por tres propiedades fundamentales (Corcuera, IntroPython):

- **Identidad**: un identificador único en memoria (obtenible con `id()`).
- **Tipo**: la clase del objeto (obtenible con `type()`).
- **Valor**: el dato que almacena.

### Tipado Dinámico

Python tiene **tipado dinámico**, lo que significa que:
- El tipo está asociado al **valor (objeto)** en memoria, **no a la variable** en sí.
- No necesitas declarar el tipo de una variable antes de usarla.
- Una variable se crea automáticamente cuando un valor es asignado la primera vez.
- Una variable puede reasignarse a objetos de cualquier tipo.

> **Puente con C:** A diferencia de C, donde una variable reserva un espacio de memoria con un tamaño fijo para un tipo específico (`int x = 5;`), en Python **todas las variables actúan como punteros (referencias)** a objetos en memoria. No existen tipos primitivos "por valor". El recolector de basura (Garbage Collector) se encarga de liberar la memoria cuando un objeto ya no tiene referencias.

> **Concepto clave (MoureDev, Guía Junior 2026):** Python usa **tipado dinámico** (tipos verificados durante la ejecución) y **tipado fuerte** (conversiones de tipo deben ser explícitas). Esto contrasta con C que es de tipado estático (tipos verificados en compilación).

```python
# Ejemplo de tipado dinámico, los errores se detectan en tiempo de ejecución (runtime)
x = 1
print(f"x = {x}, \t\ttipo = {type(x)}") # \t es un tabulador para alinear la salida

x = 1.0
print(f"x = {x}, \ttipo = {type(x)}")

x = 'hello'
print(f"x = {x}, \ttipo = {type(x)}")

x = [1, 2, 3]
print(f"x = {x}, \ttipo = {type(x)}")

# Ejemplo de tipado fuerte, el complilador o el intérprete no permite operaciones entre tipos incompatibles
# x = 1 + '1'  # Esto generará un error de tipo (TypeError) en tiempo de ejecución
```

### 4. Identidad de objetos: `id()` y el operador `is`

En Python, cada objeto tiene **tres propiedades fundamentales** (Corcuera, IntroPython):
- **Identidad:** un identificador único en memoria (su "dirección"). Se obtiene con `id()`.
- **Tipo:** la clase del objeto. Se obtiene con `type()`.
- **Valor:** el dato que almacena.

Hasta ahora hemos visto `type()` para examinar el tipo. Ahora nos enfocaremos en la **identidad** y cómo compararla correctamente.

#### `id()`: la dirección en memoria

`id(obj)` retorna un entero único que identifica al objeto durante su vida. En la implementación CPython, es la dirección de memoria.

```python
nota_a = 9.5
nota_b = 9.5

print(f"nota_a == nota_b: {nota_a == nota_b}")   # ¿Mismo valor?
print(f"nota_a is nota_b: {nota_a is nota_b}")   # ¿Mismo objeto?
print(f"id(nota_a): {id(nota_a)}")
print(f"id(nota_b): {id(nota_b)}")
print(f"id(nota_a) == id(nota_b): {id(nota_a) == id(nota_b)}")
```

```python
estudiantes_a = ["Ana", "Luis"]
estudiantes_b = ["Ana", "Luis"]   # Mismo contenido, diferente objeto
estudiantes_c = estudiantes_a       # Misma referencia

print("--- Comparando dos listas con el mismo contenido ---")
print(f"a == b: {estudiantes_a == estudiantes_b}")   # True: mismo valor
print(f"a is b: {estudiantes_a is estudiantes_b}")   # False: diferentes objetos
print(f"id(a): {id(estudiantes_a)}")
print(f"id(b): {id(estudiantes_b)}")

print("\n--- Comparando referencia directa ---")
print(f"a == c: {estudiantes_a == estudiantes_c}")   # True
print(f"a is c: {estudiantes_a is estudiantes_c}")   # True: mismo objeto
print(f"id(c): {id(estudiantes_c)}")
```

```python
nota = 9.5
calificacion = nota

print(f"nota: {nota}, id(nota): {id(nota)}")
print(f"calificacion: {calificacion}, id(calificacion): {id(calificacion)}")
print(f"nota is calificacion: {nota is calificacion}")
print(f"nota == calificacion: {nota == calificacion}")
```

#### Regla de oro: `==` para valores, `is` para identidad

- **`==` (igualdad):** compara si dos objetos tienen el **mismo valor**. Lo que normalmente quieres.
- **`is` (identidad):** compara si dos variables apuntan al **mismo objeto en memoria**. Más estricto y raro.

**¿Cuándo usar `is`?** Casi siempre con estos tres valores especiales (Corcuera, IntroPython):
- `None` (ausencia de valor)
- `True` y `False` (valores booleanos)

**¿Cuándo NO usar `is`?** Para comparar valores como números, strings o listas. Aunque a veces "funcione" por optimización interna de Python (pequeño caché de enteros), **no es garantía** y el código será frágil.

```python
promedio = None
aprobado = False

print(f"promedio is None: {promedio is None}")          # ✓ Correcto
print(f"promedio == None: {promedio == None}")          # Funciona, pero NO idiomático
print(f"aprobado is False: {aprobado is False}")        # ✓ Correcto
```

```python
nota = 9.5

if type(nota) is int:
    print("Es un entero")
elif type(nota) is float:
    print("Es un flotante")
elif type(nota) is str:
    print("Es una cadena")

print(f"Tipo: {type(nota)}")
```

## Resumen de la clase

En esta clase aprendimos los **3 pilares fundacionales** de Python antes de entrar en tipos y keywords:

### 1. Sintaxis: archivos `.py` y punto de entrada
- Los scripts se guardan con extensión **`.py`** y se ejecutan con el intérprete.
- La convención `if __name__ == '__main__':` permite distinguir entre:
  - **Ejecutar el archivo directamente** → ejecuta el bloque `main()`.
  - **Importar el archivo como módulo** → expone sus funciones sin ejecutar nada.
- Es la forma idiomática de organizar scripts reutilizables.

### 2. La función `print()`
- Es la herramienta básica para mostrar información en pantalla.
- Parámetros útiles que controlan el formato:
  - `sep`: separador entre valores (default: `" "`).
  - `end`: terminador de la línea (default: `"\n"`).
- Puede redirigir su salida con `file=` (útil para escribir a archivos o buffers).

### 3. Tipado dinámico y modelo de objetos
- **Todo en Python es un objeto:** números, strings, funciones, clases, incluso `None`.
- Las variables son **referencias** (punteros) a objetos en memoria, no contenedores.
- El tipo está asociado al **objeto**, no a la variable (tipado dinámico).
- `type(x)` nos dice qué tipo tiene el valor; `None` es el único valor del tipo `NoneType`.

### 4. Identidad de objetos: `id()` e `is`
- Cada objeto tiene una **identidad única** (su dirección en memoria).
- `id(obj)` retorna esa identidad como un entero.
- `x is y` retorna `True` solo si **ambos apuntan al mismo objeto** (no solo al mismo valor).
- **Regla:** usa `==` para comparar **valores**, usa `is` solo para comparar con `None`, `True`, `False` o cuando verifiques aliasing.
