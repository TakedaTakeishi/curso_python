## 1. ¿Qué son los módulos?

Un **módulo** es simplemente un archivo `.py` que contiene código Python (funciones, clases, variables). Los módulos nos permiten:

- **Organizar** el código en archivos lógicos
- **Reutilizar** código entre diferentes proyectos
- **Evitar** nombres duplicados (cada módulo tiene su namespace)


````python
# Importar un módulo de la biblioteca estándar
import math
print(f"Pi: {math.pi}")
print(f"Raíz cuadrada de 16: {math.sqrt(16)}")
print(f"Coseno de 0: {math.cos(0)}")
````

````python
# El módulo se carga una sola vez y se cachea
import math
print(f"ID del módulo math: {id(math)}")

# Importar de nuevo no recarga el módulo
import math
print(f"ID del módulo math (segunda vez): {id(math)}")
````

## 3. Crear tus propios módulos

Para crear un módulo, simplemente crea un archivo `.py` con funciones o variables.

**Ejemplo: `utilidades.py`**

```python
# utilidades.py
PI = 3.14159

def saludar(nombre):
    return f"Hola, {nombre}!"

def calcular_area_circulo(radio):
    return PI * radio ** 2
```


````python
# Simulando la importación de un módulo personalizado
# (En un archivo real, usarías: from utilidades import saludar)

# Como estamos en un notebook, definimos el módulo directamente
import types

modulo_utilidades = types.ModuleType('utilidades')
modulo_utilidades.PI = 3.14159
modulo_utilidades.saludar = lambda nombre: f"Hola, {nombre}!"
modulo_utilidades.calcular_area_circulo = lambda r: 3.14159 * r ** 2

# Registrar el módulo
import sys
sys.modules['utilidades'] = modulo_utilidades

# Ahora podemos importarlo
import utilidades
print(utilidades.saludar("Python"))
print(f"Área: {utilidades.calcular_area_circulo(5):.2f}")
````

## 5. Namespace y encapsulamiento

### `__name__` y `__main__`


````python
# __name__ es el nombre del módulo actual
# Cuando se ejecuta directamente: __name__ == '__main__'
# Cuando se importa: __name__ == nombre_del_módulo

print(f"__name__ actual: {__name__}")

# Patrón estándar para código ejecutable
if __name__ == '__main__':
    print("Este código se ejecuta solo si el archivo se ejecuta directamente")
````

### Variables privadas (convención)


````python
# Ejemplo de convención: _ al inicio = "privado"
# No hay privacidad real en Python, pero es una convención

class MiModulo:
    PUBLICO = "Todos pueden verme"
    _PRIVADO = "Convención: solo uso interno"
    __REALMENTE_PRIVADO = "Name mangling: _MiModulo__REALMENTE_PRIVADO"

# Acceder a las variables
obj = MiModulo()
print(f"Público: {obj.PUBLICO}")
print(f"Privado (convención): {obj._PRIVADO}")

# Acceso al name-mangled
print(f"Realmente privado: {obj._MiModulo__REALMENTE_PRIVADO}")
````

## 6. Gestión de dependencias con UV

**UV** es el gestor de paquetes moderno para Python. Maneja:

- **Dependencias** de tu proyecto
- **Entornos virtuales** aislados
- **Versiones** específicas de paquetes
- **Lock files** para reproducibilidad

### Comandos básicos

```bash
# Crear un proyecto
uv init mi_proyecto
cd mi_proyecto

# Crear entorno virtual
uv venv

# Agregar una dependencia
uv add requests

# Instalar todas las dependencias
uv sync

# Ejecutar un script en el entorno
uv run python script.py

# Ver dependencias instaladas
uv pip list
```

````python
# json: leer/escribir JSON
import json

datos = {
    "nombre": "Ana",
    "edad": 25,
    "lenguajes": ["Python", "JavaScript", "Rust"]
}

# Convertir a string JSON
json_str = json.dumps(datos, indent=2)
print("Como JSON:")
print(json_str)

# Convertir de vuelta a Python
datos_recuperados = json.loads(json_str)
print(f"\nRecuperado: {datos_recuperados['nombre']}")
````

````python
# collections: estructuras de datos avanzadas
from collections import Counter, defaultdict, namedtuple

# Counter: contar elementos
frase = "python es un lenguaje de programación python es genial"
contador = Counter(frase.split())
print(f"Palabras más comunes: {contador.most_common(3)}")

# defaultdict: diccionario con valores por defecto
grupos = defaultdict(list)
nombres = [("Ana", "ingenieros"), ("Luis", "diseñadores"), ("María", "ingenieros")]
for nombre, grupo in nombres:
    grupos[grupo].append(nombre)
print(f"\nGrupos: {dict(grupos)}")
````

## 8. Ejercicio Práctico: Crear un Paquete

**Contexto:** Vas a crear un paquete `mi_libreria` con varios módulos de utilidades estadísticas y geométricas.

**Instrucciones:**

Crea la siguiente estructura de archivos:

```
mi_libreria/
├── __init__.py
├── estadisticas.py
└── geometria.py
```

**`mi_libreria/estadisticas.py`** debe tener:
- `media(numeros)` — Calcula la media.
- `mediana(numeros)` — Calcula la mediana.
- `desviacion_estandar(numeros)` — Calcula la desviación estándar.

**`mi_libreria/geometria.py`** debe tener:
- `area_circulo(radio)` — Área de un círculo.
- `area_rectangulo(base, altura)` — Área de un rectángulo.
- `distancia_puntos(p1, p2)` — Distancia euclidiana entre dos puntos.

**`mi_libreria/__init__.py`** debe importar las funciones principales para que estén disponibles directamente.

**Pistas:**
- Usa solo el módulo `math` (no `numpy`).
- Para la desviación estándar: `sqrt(sum((x - media)²) / n)`.

**Restricción del ejercicio:**
- No uses librerías externas como `numpy`.
- Solo el módulo `math` de la biblioteca estándar.


````python
"""
Solución al Ejercicio Práctico
Estos códigos deben ir en archivos separados dentro de la carpeta mi_libreria/
"""

# mi_libreria/__init__.py
from .estadisticas import media, mediana, desviacion_estandar
from .geometria import area_circulo, area_rectangulo, distancia_puntos

__all__ = [
    'media', 'mediana', 'desviacion_estandar',
    'area_circulo', 'area_rectangulo', 'distancia_puntos'
]

# mi_libreria/estadisticas.py
import math

def media(numeros):
    """Calcula la media de una lista de números."""
    if not numeros:
        return 0
    return sum(numeros) / len(numeros)

def mediana(numeros):
    """Calcula la mediana de una lista de números."""
    if not numeros:
        return 0
    sorted_nums = sorted(numeros)
    n = len(sorted_nums)
    if n % 2 == 0:
        return (sorted_nums[n // 2 - 1] + sorted_nums[n // 2]) / 2
    return sorted_nums[n // 2]

def desviacion_estandar(numeros):
    """Calcula la desviación estándar."""
    if len(numeros) < 2:
        return 0
    med = media(numeros)
    varianza = sum((x - med) ** 2 for x in numeros) / len(numeros)
    return math.sqrt(varianza)

# mi_libreria/geometria.py
import math

def area_circulo(radio):
    """Calcula el área de un círculo."""
    return math.pi * radio ** 2

def area_rectangulo(base, altura):
    """Calcula el área de un rectángulo."""
    return base * altura

def distancia_puntos(p1, p2):
    """Calcula la distancia euclidiana entre dos puntos."""
    return math.sqrt((p2[0] - p1[0]) ** 2 + (p2[1] - p1[1]) ** 2)
````

````python
# Probar las funciones (simulando el paquete)
datos = [4, 8, 15, 16, 23, 42]
print(f"Datos: {datos}")
print(f"Media: {media(datos):.2f}")
print(f"Mediana: {mediana(datos)}")
print(f"Desviación estándar: {desviacion_estandar(datos):.2f}")

print(f"\nÁrea círculo (r=5): {area_circulo(5):.2f}")
print(f"Área rectángulo (4x3): {area_rectangulo(4, 3)}")
print(f"Distancia (0,0) a (3,4): {distancia_puntos((0, 0), (3, 4))}")
````
