## 1. Condicionales: `if`, `elif`, `else`

Las condicionales permiten ejecutar diferentes bloques de cĂłdigo segĂşn si una condiciĂłn es verdadera o falsa.

### Sintaxis bĂĄsica

```python
if condiciĂłn:
    # cĂłdigo si la condiciĂłn es True
elif otra_condiciĂłn:
    # cĂłdigo si la otra condiciĂłn es True
else:
    # cĂłdigo si ninguna condiciĂłn fue True
```

### Ejemplo simple

```python
edad = 18
if edad >= 18:
    print("Eres mayor de edad")
else:
    print("Eres menor de edad")
```


````python
edad = 20
if edad >= 18:
    print(f"Tienes {edad} aĂąos: eres mayor de edad")
else:
    print(f"Tienes {edad} aĂąos: eres menor de edad")
````

````python
# MĂşltiples condiciones con elif
nota = 8.5
if nota >= 9:
    calificacion = "A"
elif nota >= 8:
    calificacion = "B"
elif nota >= 7:
    calificacion = "C"
elif nota >= 6:
    calificacion = "D"
else:
    calificacion = "F"

print(f"Nota: {nota} â CalificaciĂłn: {calificacion}")
````

````python
# Operadores de comparaciĂłn encadenados (exclusivo de Python)
edad = 25
if 18 <= edad < 65:
    print(f"Edad {edad}: adulto en edad laboral")
else:
    print(f"Edad {edad}: fuera del rango laboral tĂ­pico")
````

## 2. Bucle `for`

El bucle `for` en Python itera sobre una secuencia (lista, string, range, etc.). A diferencia de C, **no es un contador**, es un iterador.


````python
# Iterar sobre una lista
frutas = ["manzana", "pera", "uva"]
for fruta in frutas:
    print(fruta)
````

````python
# range(): generar secuencias de nĂşmeros
# range(inicio, fin, paso) - el fin es exclusivo
for i in range(5):  # 0, 1, 2, 3, 4
    print(f"IteraciĂłn {i}")
````

````python
# range con inicio y fin
for i in range(2, 10):  # 2, 3, 4, 5, 6, 7, 8, 9
    print(i, end=" ")
print()
````

````python
# range con paso
for i in range(0, 20, 3):  # 0, 3, 6, 9, 12, 15, 18
    print(i, end=" ")
print()
````

````python
# enumerate(): obtener Ă­ndice y valor
colores = ["rojo", "verde", "azul"]
for indice, color in enumerate(colores):
    print(f"  {indice}: {color}")
````

````python
# zip(): iterar sobre mĂşltiples listas en paralelo
nombres = ["Ana", "Luis", "MarĂ­a"]
edades = [25, 30, 28]
ciudades = ["CDMX", "Madrid", "BogotĂĄ"]

for nombre, edad, ciudad in zip(nombres, edades, ciudades):
    print(f"{nombre} ({edad} aĂąos) vive en {ciudad}")
````

## 3. Bucle `while`

El bucle `while` repite un bloque mientras una condiciĂłn sea verdadera. Ăsalo cuando **no sabes cuĂĄntas iteraciones** necesitas de antemano.


````python
# While bĂĄsico: contar hasta 5
contador = 1
while contador <= 5:
    print(contador)
    contador += 1
````

````python
# While con condiciĂłn dinĂĄmica
# Simular crecimiento de inversiĂłn
capital = 1000
tasa = 0.05  # 5% anual
aĂąos = 0
while capital < 2000:
    capital *= (1 + tasa)
    aĂąos += 1

print(f"Se necesitaron {aĂąos} aĂąos para duplicar la inversiĂłn")
print(f"Capital final: ${capital:.2f}")
````

## 4. Control de bucles: `break`, `continue`, `pass`

### `break` â Salir del bucle inmediatamente


````python
# Buscar el primer nĂşmero par en una lista
numeros = [1, 3, 5, 8, 9, 11]
for num in numeros:
    if num % 2 == 0:
        print(f"Primer par encontrado: {num}")
        break
else:
    print("No se encontraron pares")
````

````python
# El bucle for tiene un 'else' que se ejecuta si NO hubo break
# Esto es exclusivo de Python
for i in range(1, 6):
    if i == 10:  # Nunca se cumple
        break
else:
    print("El bucle terminĂł sin break")
````

### `continue` â Saltar a la siguiente iteraciĂłn


````python
# Imprimir solo nĂşmeros impares
for i in range(1, 11):
    if i % 2 == 0:
        continue  # Saltar pares
    print(i, end=" ")
print()
````

````python
# pass es Ăştil para clases o funciones vacĂ­as
class MiClaseVacia:
    pass

def funcion_pendiente():
    pass  # TODO: implementar mĂĄs tarde

print(f"Clase vacĂ­a: {MiClaseVacia()}")
print(f"FunciĂłn pendiente: {funcion_pendiente()}")
````

## 6. Operadores lĂłgicos: `and`, `or`, `not`

Combinan condiciones para formar reglas mĂĄs complejas.

### `and` â Ambas condiciones deben ser True


````python
edad = 25
tiene_licencia = True
if edad >= 18 and tiene_licencia:
    print("Puede conducir")
else:
    print("No puede conducir")
````

````python
# Cortocircuito: si la primera es False, la segunda no se evalĂşa
x = 0
if x != 0 and 10 / x > 2:  # Seguro: no divide por cero
    print("Resultado")
else:
    print("No se evalĂşa la divisiĂłn (cortocircuito)")
````

### `or` â Al menos una condiciĂłn debe ser True


````python
dia = "sĂĄbado"
if dia == "sĂĄbado" or dia == "domingo":
    print("Es fin de semana")
else:
    print("Es dĂ­a laboral")
````

### `not` â Invierte el valor lĂłgico


````python
es_mayor = False
if not es_mayor:
    print("No es mayor de edad")
else:
    print("Es mayor de edad")
````

````python
# CombinaciĂłn de operadores lĂłgicos
edad = 30
tiene_trabajo = True
tiene_credito = True

apto_credito = (edad >= 18 and edad <= 65) and (tiene_trabajo or tiene_credito)
print(f"ÂżApto para crĂŠdito? {apto_credito}")
````

## 7. Ejercicio PrĂĄctico: Validador de ContraseĂąas Seguras

**Contexto:** Vas a crear un validador que verifique si una contraseĂąa cumple criterios de seguridad: longitud mĂ­nima, mayĂşsculas, minĂşsculas, nĂşmeros y caracteres especiales.

**Instrucciones:**

Crea una funciĂłn `validar_contraseĂąa(contraseĂąa)` que retorne `True` si cumple TODOS estos criterios:
- MĂ­nimo 8 caracteres
- Al menos una letra mayĂşscula
- Al menos una letra minĂşscula
- Al menos un dĂ­gito
- Al menos un carĂĄcter especial (de `!@#$%^&*`)

Luego, crea un programa que use esta funciĂłn para validar una contraseĂąa ingresada por el usuario.

**Pistas:**
- Usa bucles `for` para iterar sobre cada carĂĄcter
- Usa variables de bandera (flag) para verificar cada criterio
- No uses expresiones regulares (`re`); solo mĂŠtodos de strings y bucles

**RestricciĂłn del ejercicio:**
- No se pueden usar expresiones regulares.
- No uses listas todavĂ­a; solo variables simples y bucles.


````python
"""
SoluciĂłn al Ejercicio PrĂĄctico: Validador de ContraseĂąas
Este cĂłdigo debe ir en un archivo .py separado porque usa input().
AquĂ­ mostramos la lĂłgica sin input() para que se ejecute en el notebook.
"""

def validar_contraseĂąa(contraseĂąa):
    """Valida si una contraseĂąa cumple los criterios de seguridad."""
    # Inicializar banderas
    tiene_mayuscula = False
    tiene_minuscula = False
    tiene_digito = False
    tiene_especial = False
    longitud_valida = False
    caracteres_especiales = "!@#$%^&*"
    
    # Verificar longitud
    if len(contraseĂąa) >= 8:
        longitud_valida = True
    
    # Verificar cada carĂĄcter
    for caracter in contraseĂąa:
        if caracter.isupper():
            tiene_mayuscula = True
        elif caracter.islower():
            tiene_minuscula = True
        elif caracter.isdigit():
            tiene_digito = True
        elif caracter in caracteres_especiales:
            tiene_especial = True
    
    # La contraseĂąa es vĂĄlida si TODOS los criterios se cumplen
    es_valida = (longitud_valida and tiene_mayuscula and 
                 tiene_minuscula and tiene_digito and tiene_especial)
    
    return es_valida, {
        "longitud": longitud_valida,
        "mayuscula": tiene_mayuscula,
        "minuscula": tiene_minuscula,
        "digito": tiene_digito,
        "especial": tiene_especial
    }
````

````python
# Probar el validador con diferentes contraseĂąas
contraseĂąas_test = [
    "abc",           # Muy corta
    "abcdefgh",      # Sin mayĂşscula, dĂ­gito ni especial
    "Abcdefgh",      # Sin dĂ­gito ni especial
    "Abcdefg1",      # Sin especial
    "Abcdef1!",      # VĂĄlida
    "Xyz12345@",     # VĂĄlida
]

for contraseĂąa in contraseĂąas_test:
    es_valida, detalles = validar_contraseĂąa(contraseĂąa)
    estado = "â VĂLIDA" if es_valida else "â INVĂLIDA"
    print(f"{estado} '{contraseĂąa}': {detalles}")
    print()
````
