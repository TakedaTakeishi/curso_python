## 1. Condicionales: `if`, `elif`, `else`

Las condicionales permiten ejecutar diferentes bloques de código según si una condición es verdadera o falsa.

### Sintaxis básica

```python
if condición:
    # código si la condición es True
elif otra_condición:
    # código si la otra condición es True
else:
    # código si ninguna condición fue True
```

### Ejemplo simple

```python
edad = 18
if edad >= 18:
    print("Eres mayor de edad")
else:
    print("Eres menor de edad")
```


````python
edad = 20
if edad >= 18:
    print(f"Tienes {edad} años: eres mayor de edad")
else:
    print(f"Tienes {edad} años: eres menor de edad")
````

````python
# Múltiples condiciones con elif
nota = 8.5
if nota >= 9:
    calificacion = "A"
elif nota >= 8:
    calificacion = "B"
elif nota >= 7:
    calificacion = "C"
elif nota >= 6:
    calificacion = "D"
else:
    calificacion = "F"

print(f"Nota: {nota} → Calificación: {calificacion}")
````

````python
# Operadores de comparación encadenados (exclusivo de Python)
edad = 25
if 18 <= edad < 65:
    print(f"Edad {edad}: adulto en edad laboral")
else:
    print(f"Edad {edad}: fuera del rango laboral típico")
````

## 1.5 `match` / `case` (Python 3.10+): pattern matching

Desde **Python 3.10** existe el **structural pattern matching** con `match`/`case`. Es el equivalente de `switch`/`case` en otros lenguajes, pero mucho más poderoso porque puede coincidir con **estructuras** (no solo valores).

### Sintaxis básica

```python
match valor:
    case patron1:
        # código si coincide con patron1
    case patron2:
        # código si coincide con patron2
    case _:
        # caso por defecto (como else)
```

### Ejemplo básico con valores

````python
dia = "lunes"

match dia:
    case "lunes" | "martes" | "miércoles" | "jueves" | "viernes":
        print(f"{dia} es día laboral")
    case "sábado" | "domingo":
        print(f"{dia} es fin de semana")
    case _:
        print(f"¿{dia}? No reconozco ese día")

# El operador | permite múltiples patrones en un mismo case
````

````python
# Comparación con if/elif/else equivalente
# match/case es más legible cuando hay muchos casos sobre el mismo valor

def clasificar_edad(edad):
    match edad:
        case 0 | 1 | 2:
            return "bebé"
        case 3 | 4 | 5:
            return "preescolar"
        case 6 | 7 | 8 | 9 | 10 | 11:
            return "niño"
        case 12 | 13 | 14 | 15 | 16 | 17:
            return "adolescente"
        case _ if edad >= 18 and edad < 65:
            return "adulto"
        case _ if edad >= 65:
            return "adulto mayor"
        case _:
            return "edad inválida"

print(f"5 años → {clasificar_edad(5)}")
print(f"20 años → {clasificar_edad(20)}")
print(f"70 años → {clasificar_edad(70)}")

# Nota: case _ if condicion es "guarda" (pattern guard)
````

### Captura de variables con `as`

`match`/`case` puede **desempaquetar** estructuras y capturar valores:

````python
# Capturar partes de una tupla
punto = (3, 4)

match punto:
    case (0, 0):
        print("Origen")
    case (x, 0):
        print(f"Eje X, en x={x}")
    case (0, y):
        print(f"Eje Y, en y={y}")
    case (x, y) if x == y:
        print(f"Diagonal, en ({x}, {y})")
    case (x, y):
        print(f"Punto general: ({x}, {y})")

# Capturar en diccionarios
comando = {"accion": "buscar", "query": "python"}

match comando:
    case {"accion": "buscar", "query": q}:
        print(f"Buscando: {q}")
    case {"accion": "borrar", "id": id}:
        print(f"Borrando ID: {id}")
    case {"accion": accion}:
        print(f"Acción no reconocida: {accion}")
````

### Cuándo usar `match`/`case` vs `if`/`elif`/`else`

| Escenario | Usa |
|---|---|
| Muchas condiciones sobre el **mismo valor** | `match`/`case` (más legible) |
| Condiciones sobre **valores diferentes** | `if`/`elif`/`else` (más natural) |
| Coincidir con **estructuras** (tuplas, dicts) | `match`/`case` (más poderoso) |
| Condiciones **complejas** (rangos, comparaciones) | `if`/`elif`/`else` con guards en match |
| **Compatibilidad** con Python < 3.10 | `if`/`elif`/`else` (única opción) |

> **Nota importante:** `match`/`case` **no es un reemplazo** de `if`/`elif`/`else`. Es una **herramienta complementaria** que brilla en casos específicos (muchos casos sobre el mismo valor, coincidencia de estructuras).

## 2. Bucle `for`

El bucle `for` en Python itera sobre una secuencia (lista, string, range, etc.). A diferencia de C, **no es un contador**, es un iterador. Esto quiere decir que extrae un elemento de cualquier objeto que sea un iterador, lo asigna a una variabe y lo introduce dentro del bloque. De manera visual sería:

### Sintaxis básica

```python
for variable in iterador:
    # Bloque
    print(variable)
``` 

### Ejemplos
A continuación se pueden ejecutar algunos ejemplos de utilización del for.

````python

# Iterar sobre una lista
frutas = ["manzana", "pera", "uva"]
for fruta in frutas:
    print(fruta)
````

````python
# range(): generar secuencias de números
# range(inicio, fin, paso) - el fin es exclusivo
for i in range(5):  # 0, 1, 2, 3, 4
    print(f"Iteración {i}")
````

````python
# range con inicio y fin
for i in range(2, 10):  # 2, 3, 4, 5, 6, 7, 8, 9
    print(i, end=" ")
print()
````

````python
# range con paso
for i in range(0, 20, 3):  # 0, 3, 6, 9, 12, 15, 18
    print(i, end=" ")
print()
````

````python
# enumerate(): obtener índice y valor
colores = ["rojo", "verde", "azul"]
for indice, color in enumerate(colores):
    print(f"  {indice}: {color}")
````

````python
# zip(): iterar sobre múltiples listas en paralelo
nombres = ["Ana", "Luis", "María"]
edades = [25, 30, 28]
ciudades = ["CDMX", "Madrid", "Bogotá"]

for nombre, edad, ciudad in zip(nombres, edades, ciudades):
    print(f"{nombre} ({edad} años) vive en {ciudad}")
````

## 3. Bucle `while`

El bucle `while` repite un bloque mientras una condición sea verdadera. Úsalo cuando **no sabes cuántas iteraciones** necesitas de antemano.


````python
# While básico: contar hasta 5
contador = 1
while contador <= 5:
    print(contador)
    contador += 1
````

````python
# While con condición dinámica
# Simular crecimiento de inversión
capital = 1000
tasa = 0.05  # 5% anual
años = 0
while capital < 2000:
    capital *= (1 + tasa)
    años += 1

print(f"Se necesitaron {años} años para duplicar la inversión")
print(f"Capital final: ${capital:.2f}")
````

## 4. Control de bucles: `break`, `continue`, `pass`

### `break` — Salir del bucle inmediatamente


````python
# Buscar el primer número par en una lista
numeros = [1, 3, 5, 8, 9, 11]
for num in numeros:
    if num % 2 == 0:
        print(f"Primer par encontrado: {num}")
        break
else:
    print("No se encontraron pares")
````

````python
# El bucle for tiene un 'else' que se ejecuta si NO hubo break
# Esto es exclusivo de Python
for i in range(1, 6):
    if i == 10:  # Nunca se cumple
        break
else:
    print("El bucle terminó sin break")
````

### `continue` — Saltar a la siguiente iteración


````python
# Imprimir solo números impares
for i in range(1, 11):
    if i % 2 == 0:
        continue  # Saltar pares
    print(i, end=" ")
print()
````

````python
# pass es útil para clases o funciones vacías
class MiClaseVacia:
    pass

def funcion_pendiente():
    pass  # TODO: implementar más tarde

print(f"Clase vacía: {MiClaseVacia()}")
print(f"Función pendiente: {funcion_pendiente()}")
````

## 5. Operadores `in` y `not in`

Verifican si un elemento está (o no) en una secuencia.


````python
# Membresía en listas
frutas = ["manzana", "pera", "uva"]
print(f"¿'manzana' in frutas? {'manzana' in frutas}")
print(f"¿'naranja' not in frutas? {'naranja' not in frutas}")
````

````python
# Membresía en strings
mensaje = "Hola mundo"
print(f"¿'Hola' in mensaje? {'Hola' in mensaje}")
print(f"¿'Python' in mensaje? {'Python' in mensaje}")
````

## 6. Operadores lógicos: `and`, `or`, `not`

Combinan condiciones para formar reglas más complejas.

### `and` — Ambas condiciones deben ser True


````python
edad = 25
tiene_licencia = True
if edad >= 18 and tiene_licencia:
    print("Puede conducir")
else:
    print("No puede conducir")
````

````python
# Cortocircuito: si la primera es False, la segunda no se evalúa
x = 0
if x != 0 and 10 / x > 2:  # Seguro: no divide por cero
    print("Resultado")
else:
    print("No se evalúa la división (cortocircuito)")
````

### `or` — Al menos una condición debe ser True


````python
dia = "sábado"
if dia == "sábado" or dia == "domingo":
    print("Es fin de semana")
else:
    print("Es día laboral")
````

### `not` — Invierte el valor lógico


````python
es_mayor = False
if not es_mayor:
    print("No es mayor de edad")
else:
    print("Es mayor de edad")
````

````python
# Combinación de operadores lógicos
edad = 30
tiene_trabajo = True
tiene_credito = True

apto_credito = (edad >= 18 and edad <= 65) and (tiene_trabajo or tiene_credito)
print(f"¿Apto para crédito? {apto_credito}")
````

## 7. Ejercicio Práctico: Validador de Contraseñas Seguras

**Contexto:** Vas a crear un validador que verifique si una contraseña cumple criterios de seguridad: longitud mínima, mayúsculas, minúsculas, números y caracteres especiales.

**Instrucciones:**

Crea una función `validar_contraseña(contraseña)` que retorne `True` si cumple TODOS estos criterios:
- Mínimo 8 caracteres
- Al menos una letra mayúscula
- Al menos una letra minúscula
- Al menos un dígito
- Al menos un carácter especial (de `!@#$%^&*`)

Luego, crea un programa que use esta función para validar una contraseña ingresada por el usuario.

**Pistas:**
- Usa bucles `for` para iterar sobre cada carácter
- Usa variables de bandera (flag) para verificar cada criterio
- No uses expresiones regulares (`re`); solo métodos de strings y bucles

**Restricción del ejercicio:**
- No se pueden usar expresiones regulares.
- No uses listas todavía; solo variables simples y bucles.


````python
"""
Solución al Ejercicio Práctico: Validador de Contraseñas
Este código debe ir en un archivo .py separado porque usa input().
Aquí mostramos la lógica sin input() para que se ejecute en el notebook.
"""

def validar_contraseña(contraseña):
    """Valida si una contraseña cumple los criterios de seguridad."""
    # Inicializar banderas
    tiene_mayuscula = False
    tiene_minuscula = False
    tiene_digito = False
    tiene_especial = False
    longitud_valida = False
    caracteres_especiales = "!@#$%^&*"
    
    # Verificar longitud
    if len(contraseña) >= 8:
        longitud_valida = True
    
    # Verificar cada carácter
    for caracter in contraseña:
        if caracter.isupper():
            tiene_mayuscula = True
        elif caracter.islower():
            tiene_minuscula = True
        elif caracter.isdigit():
            tiene_digito = True
        elif caracter in caracteres_especiales:
            tiene_especial = True
    
    # La contraseña es válida si TODOS los criterios se cumplen
    es_valida = (longitud_valida and tiene_mayuscula and 
                 tiene_minuscula and tiene_digito and tiene_especial)
    
    return es_valida, {
        "longitud": longitud_valida,
        "mayuscula": tiene_mayuscula,
        "minuscula": tiene_minuscula,
        "digito": tiene_digito,
        "especial": tiene_especial
    }
````

````python
# Probar el validador con diferentes contraseñas
contraseñas_test = [
    "abc",           # Muy corta
    "abcdefgh",      # Sin mayúscula, dígito ni especial
    "Abcdefgh",      # Sin dígito ni especial
    "Abcdefg1",      # Sin especial
    "Abcdef1!",      # Válida
    "Xyz12345@",     # Válida
]

for contraseña in contraseñas_test:
    es_valida, detalles = validar_contraseña(contraseña)
    estado = "✓ VÁLIDA" if es_valida else "✗ INVÁLIDA"
    print(f"{estado} '{contraseña}': {detalles}")
    print()
````
