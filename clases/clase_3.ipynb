## 1. ¿Qué son las palabras reservadas?

Al igual que en C, Python tiene **palabras clave** (keywords) que no puedes usar como identificadores (nombres de variables, funciones, clases, etc.). Estas palabras tienen un significado especial para el intérprete y son la base de la sintaxis del lenguaje.

```python
# Esto es un SyntaxError
for = 5  # ❌ No puedes usar 'for' como nombre de variable
```

En Python 3 hay **35 palabras reservadas**. Aunque muchas las estudiaremos en profundidad más adelante, conviene tenerlas presentes desde el inicio.


## 2. Las 35 palabras reservadas

| | | | | | |
|---|---|---|---|---|---|
| `False` | `None` | `True` | `and` | `as` | `assert` |
| `async` | `await` | `break` | `class` | `continue` | `def` |
| `del` | `elif` | `else` | `except` | `finally` | `for` |
| `from` | `global` | `if` | `import` | `in` | `is` |
| `lambda` | `nonlocal` | `not` | `or` | `pass` | `raise` |
| `return` | `try` | `while` | `with` | `yield` | |

Puedes obtener la lista desde el módulo `keyword`:


````python
import keyword
print(f"Total de palabras reservadas: {len(keyword.kwlist)}")
print(keyword.kwlist)
````

````python
# Verificar si una palabra es reservada
print(f"¿'for' es reservada? {keyword.iskeyword('for')}")
print(f"¿'variable' es reservada? {keyword.iskeyword('variable')}")
print(f"¿'class' es reservada? {keyword.iskeyword('class')}")
````

## 3. El operador asignación (`=`)

En Python **no es necesario declarar las variables** antes de usarlas. La asignación inicial crea la variable y enlaza el valor.

Python soporta varias formas de asignación:

- **Asignación simple:** `x = 5`
- **Asignación múltiple (por pares):** `b, c, d = 123, 4.5, 'Hello'`
- **Asignación en cadena:** `e = f = g = 123`
- **Eliminación:** `del x` elimina la variable


````python
# Asignación simple
x = 5
print(f"x = {x}")
````

````python
# Asignación múltiple
b, c, d = 123, 4.5, 'Hello'
print(f"b = {b}, c = {c}, d = {d}")
````

````python
# Asignación en cadena
e = f = g = 123
print(f"e = {e}, f = {f}, g = {g}")
````

````python
# Eliminación con del
x = 10
print(f"Antes: x = {x}")
del x
# print(x)  # NameError: name 'x' is not defined
print("x fue eliminada")
````

## 4. Importación de módulos

Una de las keywords más importantes es `import`, que nos permite usar código de otros archivos o de la biblioteca estándar.

### `import` — Importar un módulo completo

```python
import math
print(math.pi)
```

### `from ... import` — Importar elementos específicos

```python
from math import pi, sqrt
print(pi)
```

### `as` — Crear un alias

```python
import numpy as np
```


````python
import math
print(f"Pi: {math.pi}")
print(f"Raíz cuadrada de 16: {math.sqrt(16)}")
````

````python
from math import pi, sqrt
print(f"Pi: {pi}")
print(f"Raíz cuadrada de 25: {sqrt(25)}")
````

````python
import statistics as stats
notas = [8.5, 9.0, 7.5, 10.0]
print(f"Promedio: {stats.mean(notas)}")
print(f"Mediana: {stats.median(notas)}")
````

## 5. Punto de entrada: `if __name__ == '__main__':`

Esta es una de las convenciones más importantes de Python. Permite definir un **punto de entrada** explícito, similar a `main()` en C.

### ¿Cómo funciona?

Cuando ejecutas un archivo Python directamente, Python asigna a la variable especial `__name__` el valor `"__main__"`. Pero cuando el archivo es **importado** como módulo desde otro archivo, `__name__` toma el nombre del archivo.


````python
def saludar(nombre):
    """Función que se puede importar o ejecutar directamente."""
    return f"¡Hola, {nombre}!"

def main():
    """Función principal del script."""
    nombre = input("¿Cómo te llamas? ")
    print(saludar(nombre))

# Punto de entrada
if __name__ == '__main__':
    main()
````

> **Nota:** En este notebook no podemos usar `input()` directamente, pero puedes copiar este código a un archivo `.py` y ejecutarlo con `uv run python archivo.py`.

### ¿Por qué es importante?

- **Reutilización de código:** Las funciones definidas en un archivo pueden importarse desde otros sin ejecutar el código principal.
- **Organización:** Separa la definición de funciones de su ejecución.
- **Testing:** Permite importar funciones para probarlas sin ejecutar el script.


## 6. Ejercicio Práctico: Script modular

**Contexto:** Vas a crear dos archivos Python que demuestren el uso del punto de entrada y la importación de módulos.

**Instrucciones:**

1. Crea un archivo `utilidades.py` con:
   - Una función `saludar(nombre)` que retorne un saludo.
   - Una función `calcular_promedio(notas)` que retorne el promedio de una lista.
   - Un bloque `if __name__ == '__main__':` que ejecute las funciones con datos de ejemplo.

2. Crea un archivo `cliente.py` que:
   - Importe las funciones de `utilidades.py` usando `from ... import`.
   - Las use con datos diferentes a los del ejemplo.

**Pistas:**
- `from utilidades import saludar, calcular_promedio`
- Los archivos deben estar en la misma carpeta.

**Restricción del ejercicio:**
- Solo se pueden usar palabras reservadas con su propósito; no se permite redefinir keywords ni crear variables con nombres reservados.
- No uses listas dentro de `utilidades.py` (solo tuplas o variables simples) para mantener el ejemplo simple.


````python
"""
Solución al Ejercicio Práctico
Estos códigos deben ir en archivos .py separados.

Archivo 1: utilidades.py
"""

# utilidades.py
def saludar(nombre):
    """Retorna un saludo personalizado."""
    return f"¡Hola, {nombre}!"

def calcular_promedio(notas):
    """Calcula el promedio de una lista de notas."""
    if not notas:
        return 0.0
    return sum(notas) / len(notas)

if __name__ == '__main__':
    # Esto solo se ejecuta si corres utilidades.py directamente
    print(saludar("Python"))
    print(f"Promedio: {calcular_promedio([8.5, 9.0, 7.5])}")
````

````python
"""
Archivo 2: cliente.py
Este archivo importa y usa las funciones de utilidades.py
"""

# cliente.py
from utilidades import saludar, calcular_promedio

# Usamos las funciones importadas con datos diferentes
nombre_usuario = "Estudiante"
notas_estudiante = [10.0, 9.5, 8.0, 9.0]

print(saludar(nombre_usuario))
print(f"Tu promedio es: {calcular_promedio(notas_estudiante)}")
````
