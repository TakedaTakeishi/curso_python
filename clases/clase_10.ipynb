## 1. Notación Big O

La notación **Big O** describe cómo escala el tiempo (o espacio) de un algoritmo cuando crece el tamaño de la entrada (n).

### Clases de complejidad comunes

| Notación | Nombre | Ejemplo |
|---|---|---|
| O(1) | Constante | Acceso a array por índice |
| O(log n) | Logarítmica | Búsqueda binaria |
| O(n) | Lineal | Recorrer una lista |
| O(n log n) | Linearítmica | Mergesort, Quicksort |
| O(n²) | Cuadrática | Burbuja, selección |
| O(2ⁿ) | Exponencial | Fibonacci recursivo ingenuo |


```python
import time
import math

# Demostrar el crecimiento de cada clase de complejidad
n_values = [10, 100, 1000, 10000]

print("Crecimiento de diferentes complejidades:")
print(f"{'n':>8} | {'O(1)':>8} | {'O(log n)':>10} | {'O(n)':>8} | {'O(n log n)':>12} | {'O(n²)':>10}")
print("-" * 70)
for n in n_values:
    o1 = 1
    o_log = int(math.log2(n)) if n > 0 else 0
    o_n = n
    o_nlogn = n * int(math.log2(n)) if n > 0 else 0
    o_n2 = n ** 2
    print(f"{n:>8} | {o1:>8} | {o_log:>10} | {o_n:>8} | {o_nlogn:>12} | {o_n2:>10}")
```

```python
# Ejemplo: O(1) vs O(n) vs O(n²)
def operacion_constante(lista):
    """O(1): solo accede al primer elemento."""
    return lista[0] if lista else None

def operacion_lineal(lista):
    """O(n): recorre toda la lista."""
    return sum(lista)

def operacion_cuadratica(lista):
    """O(n²): para cada elemento, recorre toda la lista."""
    contador = 0
    for x in lista:
        for y in lista:
            contador += 1
    return contador

# Comparar tiempos
lista = list(range(10_000))
inicio = time.time(); operacion_constante(lista); t1 = time.time() - inicio
inicio = time.time(); operacion_lineal(lista); t2 = time.time() - inicio
inicio = time.time(); operacion_cuadratica(lista); t3 = time.time() - inicio

print(f"O(1):   {t1*1000:.4f} ms")
print(f"O(n):   {t2*1000:.4f} ms")
print(f"O(n²):  {t3*1000:.4f} ms")
```

## 2. Búsqueda lineal y binaria

### Búsqueda lineal: O(n)


```python
def busqueda_lineal(lista, objetivo):
    """Busca un elemento recorriendo la lista. O(n)."""
    for i, elemento in enumerate(lista):
        if elemento == objetivo:
            return i
    return -1

# Probar
numeros = [4, 2, 7, 1, 9, 5, 3, 8, 6]
print(f"Lista: {numeros}")
print(f"Posición de 7: {busqueda_lineal(numeros, 7)}")
print(f"Posición de 10: {busqueda_lineal(numeros, 10)}")
```

### Búsqueda binaria: O(log n)

Requiere que la lista esté **ordenada**.


```python
def busqueda_binaria(lista, objetivo):
    """Busca dividiendo a la mitad. O(log n). Requiere lista ordenada."""
    izquierda, derecha = 0, len(lista) - 1
    
    while izquierda <= derecha:
        medio = (izquierda + derecha) // 2
        if lista[medio] == objetivo:
            return medio
        elif lista[medio] < objetivo:
            izquierda = medio + 1
        else:
            derecha = medio - 1
    return -1

# Probar
numeros_ordenados = [1, 2, 3, 4, 5, 6, 7, 8, 9]
print(f"Lista: {numeros_ordenados}")
print(f"Posición de 7: {busqueda_binaria(numeros_ordenados, 7)}")
print(f"Posición de 10: {busqueda_binaria(numeros_ordenados, 10)}")
```

```python
# Comparar rendimiento
import random
lista_grande = sorted(random.sample(range(1_000_000), 100_000))

# Búsqueda lineal
inicio = time.time()
for _ in range(1000):
    busqueda_lineal(lista_grande, 999_999)
t_lineal = time.time() - inicio

# Búsqueda binaria
inicio = time.time()
for _ in range(1000):
    busqueda_binaria(lista_grande, 999_999)
t_binaria = time.time() - inicio

print(f"1000 búsquedas en lista de 100,000 elementos:")
print(f"  Lineal:  {t_lineal*1000:.2f} ms")
print(f"  Binaria: {t_binaria*1000:.2f} ms")
print(f"  Binaria es ~{t_lineal/t_binaria:.0f}x más rápida")
```

## 3. Algoritmos de ordenación O(n²)

### Ordenamiento burbuja


```python
def ordenamiento_burbuja(lista):
    """Ordena comparando pares adyacentes. O(n²)."""
    lista = lista.copy()
    n = len(lista)
    for i in range(n):
        hubo_intercambio = False
        for j in range(n - 1 - i):
            if lista[j] > lista[j + 1]:
                lista[j], lista[j + 1] = lista[j + 1], lista[j]
                hubo_intercambio = True
        if not hubo_intercambio:  # Optimización
            break
    return lista

# Probar
datos = [64, 34, 25, 12, 22, 11, 90]
print(f"Original: {datos}")
print(f"Ordenado: {ordenamiento_burbuja(datos)}")
```

### Ordenamiento por inserción


```python
def ordenamiento_insercion(lista):
    """Inserta cada elemento en su posición correcta. O(n²)."""
    lista = lista.copy()
    for i in range(1, len(lista)):
        actual = lista[i]
        j = i - 1
        while j >= 0 and lista[j] > actual:
            lista[j + 1] = lista[j]
            j -= 1
        lista[j + 1] = actual
    return lista

datos = [64, 34, 25, 12, 22, 11, 90]
print(f"Original: {datos}")
print(f"Ordenado: {ordenamiento_insercion(datos)}")
```

### Ordenamiento por selección


```python
def ordenamiento_seleccion(lista):
    """Encuentra el mínimo y lo coloca al inicio. O(n²)."""
    lista = lista.copy()
    n = len(lista)
    for i in range(n):
        min_idx = i
        for j in range(i + 1, n):
            if lista[j] < lista[min_idx]:
                min_idx = j
        lista[i], lista[min_idx] = lista[min_idx], lista[i]
    return lista

datos = [64, 34, 25, 12, 22, 11, 90]
print(f"Original: {datos}")
print(f"Ordenado: {ordenamiento_seleccion(datos)}")
```

## 4. Quicksort y Mergesort (O(n log n))

### Quicksort: divide y vencerás


```python
def quicksort(lista):
    """Ordena dividiendo por un pivote. O(n log n) promedio."""
    if len(lista) <= 1:
        return lista
    pivote = lista[len(lista) // 2]
    menores = [x for x in lista if x < pivote]
    iguales = [x for x in lista if x == pivote]
    mayores = [x for x in lista if x > pivote]
    return quicksort(menores) + iguales + quicksort(mayores)

datos = [64, 34, 25, 12, 22, 11, 90]
print(f"Original: {datos}")
print(f"Ordenado: {quicksort(datos)}")
```

### Mergesort: divide y combina


```python
def mergesort(lista):
    """Ordena dividiendo y fusionando. O(n log n) garantizado."""
    if len(lista) <= 1:
        return lista
    
    medio = len(lista) // 2
    izquierda = mergesort(lista[:medio])
    derecha = mergesort(lista[medio:])
    
    return fusionar(izquierda, derecha)

def fusionar(izquierda, derecha):
    """Fusiona dos listas ordenadas en una sola ordenada."""
    resultado = []
    i = j = 0
    while i < len(izquierda) and j < len(derecha):
        if izquierda[i] <= derecha[j]:
            resultado.append(izquierda[i])
            i += 1
        else:
            resultado.append(derecha[j])
            j += 1
    resultado.extend(izquierda[i:])
    resultado.extend(derecha[j:])
    return resultado

datos = [64, 34, 25, 12, 22, 11, 90]
print(f"Original: {datos}")
print(f"Ordenado: {mergesort(datos)}")
```

## 5. Técnicas algorítmicas

### Divide y vencerás

Dividir el problema en subproblemas más pequeños, resolverlos recursivamente, y combinar las soluciones.

Ejemplos: quicksort, mergesort, búsqueda binaria.

### Algoritmos voraces (greedy)

En cada paso, elegir la opción que parece mejor localmente, esperando que conduzca a una solución global óptima.


```python
# Ejemplo: cambio de monedas (algoritmo voraz)
def cambio_monedas(monto, monedas_disponibles):
    """Retorna el cambio usando el menor número de monedas."""
    monedas = sorted(monedas_disponibles, reverse=True)
    resultado = []
    for moneda in monedas:
        while monto >= moneda:
            resultado.append(moneda)
            monto -= moneda
    return resultado if monto == 0 else None

# Probar
monedas = [1, 5, 10, 25, 50, 100]  # centavos
print(f"Cambio para 287 centavos: {cambio_monedas(287, monedas)}")
print(f"Cambio para 99 centavos: {cambio_monedas(99, monedas)}")
```

### Dos punteros

Usar dos índices que se mueven por la colección desde extremos opuestos o a diferentes velocidades.


```python
def suma_dos_ordenados(lista, objetivo):
    """Encuentra dos números que suman objetivo. O(n) con dos punteros."""
    izquierda, derecha = 0, len(lista) - 1
    while izquierda < derecha:
        suma = lista[izquierda] + lista[derecha]
        if suma == objetivo:
            return (lista[izquierda], lista[derecha])
        elif suma < objetivo:
            izquierda += 1
        else:
            derecha -= 1
    return None

# Probar
numeros = [1, 3, 5, 7, 9, 11, 13, 15]
print(f"Lista: {numeros}")
print(f"Pares que suman 12: {suma_dos_ordenados(numeros, 12)}")
print(f"Pares que suman 20: {suma_dos_ordenados(numeros, 20)}")
```

## 7. Ejercicio Práctico: Comparador de Algoritmos

**Contexto:** Vas a crear un script que compare el rendimiento de los algoritmos de ordenación con diferentes tamaños de entrada y muestre los resultados en una tabla.

**Instrucciones:**

Crea un archivo `comparador.py` con:

1. Implementaciones de burbuja, inserción, selección, quicksort y mergesort.
2. Una función `medir_tiempo(algoritmo, datos)` que mide cuánto tarda un algoritmo.
3. Una función `comparar_algoritmos(tamaños)` que prueba cada algoritmo con cada tamaño.
4. Una función que imprime los resultados en formato de tabla.

**Pistas:**
- Usa `time.perf_counter()` para mayor precisión.
- Genera datos aleatorios con `random.randint()`.
- Para cada tamaño, ejecuta cada algoritmo y guarda el tiempo.

**Restricción del ejercicio:**
- No uses `sorted()` ni `list.sort()` de Python.
- Implementa todos los algoritmos manualmente.


```python
"""
Solución al Ejercicio Práctico
Este código debe ir en un archivo comparador.py
"""

import time
import random

def medir_tiempo(algoritmo, datos):
    """Mide el tiempo de ejecución de un algoritmo."""
    copia = datos.copy()
    inicio = time.perf_counter()
    algoritmo(copia)
    return time.perf_counter() - inicio

def comparar_algoritmos(tamaños):
    """Compara todos los algoritmos con diferentes tamaños de entrada."""
    algoritmos = [
        ("Burbuja", ordenamiento_burbuja),
        ("Inserción", ordenamiento_insercion),
        ("Selección", ordenamiento_seleccion),
        ("Quicksort", quicksort),
        ("Mergesort", mergesort),
    ]
    
    resultados = {}
    for tamaño in tamaños:
        datos = [random.randint(1, tamaño * 10) for _ in range(tamaño)]
        resultados[tamaño] = {}
        for nombre, funcion in algoritmos:
            resultados[tamaño][nombre] = medir_tiempo(funcion, datos)
    return resultados

def imprimir_tabla(resultados):
    """Imprime los resultados en formato de tabla."""
    algoritmos = list(next(iter(resultados.values())).keys())
    
    header = f"{'Tamaño':>10} | " + " | ".join(f"{a:>10}" for a in algoritmos)
    print(header)
    print("-" * len(header))
    
    for tamaño, tiempos in resultados.items():
        fila = f"{tamaño:>10} | " + " | ".join(f"{tiempos[a]*1000:>8.2f}ms" for a in algoritmos)
        print(fila)
```

```python
# Probar el comparador
tamaños = [100, 300, 500]
print("Comparación de algoritmos de ordenación:\n")
resultados = comparar_algoritmos(tamaños)
imprimir_tabla(resultados)
```
