## 1. Diccionarios: creación y acceso

Un **diccionario** es una colección de pares **clave-valor**. Se define con llaves `{}`.


````python
# Crear diccionario
persona = {
    "nombre": "Ana",
    "edad": 25,
    "ciudad": "CDMX",
    "profesion": "Ingeniera"
}

print(persona)
print(f"Tipo: {type(persona).__name__}")
````

````python
# Acceso por clave
print(f"Nombre: {persona['nombre']}")
print(f"Edad: {persona['edad']}")

# Acceso seguro con .get() (retorna None si no existe)
print(f"Email: {persona.get('email', 'No tiene email')}")

# Acceso con [] lanza KeyError si no existe
# print(persona['email'])  # KeyError
````

````python
# Modificar valores
persona["edad"] = 26
print(f"Nueva edad: {persona['edad']}")

# Agregar nuevas claves
persona["email"] = "ana@ejemplo.com"
print(f"Con email: {persona}")
````

## 2. Métodos de diccionarios


````python
# keys(), values(), items()
persona = {"nombre": "Ana", "edad": 25, "ciudad": "CDMX"}

print("Claves:", list(persona.keys()))
print("Valores:", list(persona.values()))
print("Pares:", list(persona.items()))
````

````python
# update(): fusionar diccionarios
persona = {"nombre": "Ana", "edad": 25}
actualizacion = {"edad": 26, "ciudad": "CDMX"}
persona.update(actualizacion)
print(f"Después de update: {persona}")
````

````python
# pop(): eliminar y retornar valor
persona = {"nombre": "Ana", "edad": 25, "ciudad": "CDMX"}
edad = persona.pop("edad")
print(f"Eliminada edad: {edad}")
print(f"Diccionario: {persona}")

# pop con default
profesion = persona.pop("profesion", "Desconocida")
print(f"Profesión: {profesion}")
````

````python
# setdefault(): obtener o insertar con default
persona = {"nombre": "Ana"}
ciudad = persona.setdefault("ciudad", "CDMX")
print(f"Ciudad: {ciudad}")
print(f"Persona: {persona}")

# Si la clave ya existe, no la modifica
ciudad2 = persona.setdefault("ciudad", "Madrid")
print(f"Ciudad (existente): {ciudad2}")
````

````python
# clear(): vaciar el diccionario
temp = {"a": 1, "b": 2}
temp.clear()
print(f"Vacío: {temp}")
````

## 3. Iteración sobre diccionarios


````python
calificaciones = {"Ana": 9.5, "Luis": 7.8, "María": 8.9, "Pedro": 6.5}

# Iterar sobre claves
print("--- Iterando sobre claves ---")
for nombre in calificaciones:
    print(f"{nombre}: {calificaciones[nombre]}")
````

````python
# Iterar sobre claves y valores
print("--- Iterando sobre items ---")
for nombre, nota in calificaciones.items():
    estado = "Aprobado" if nota >= 7 else "Reprobado"
    print(f"{nombre}: {nota} ({estado})")
````

````python
# Iterar solo sobre valores
notas = list(calificaciones.values())
print(f"Notas: {notas}")
print(f"Promedio: {sum(notas) / len(notas):.2f}")
print(f"Nota más alta: {max(notas)}")
print(f"Nota más baja: {min(notas)}")
````

## 4. Sets: creación y operaciones

Un **set** es una colección **no ordenada** de elementos **únicos**. Se define con llaves `{}` (sin pares clave-valor) o con `set()`.


````python
# Crear sets
frutas = {"manzana", "pera", "uva", "manzana"}  # Los duplicados se eliminan
print(f"Frutas: {frutas}")

# Crear set vacío
vacio = set()  # ¡No usar {} porque crea un diccionario!
print(f"Vacío: {vacio}, tipo: {type(vacio).__name__}")
````

````python
# Agregar y eliminar elementos
colores = {"rojo", "verde", "azul"}
colores.add("amarillo")
print(f"Después de add: {colores}")

colores.remove("verde")
print(f"Después de remove: {colores}")

# discard no lanza error si el elemento no existe
colores.discard("rosa")
print(f"Después de discard('rosa'): {colores}")
````

````python
# Operaciones de conjuntos
a = {1, 2, 3, 4}
b = {3, 4, 5, 6}

print(f"A = {a}")
print(f"B = {b}")
print(f"Unión (A | B): {a | b}")
print(f"Intersección (A & B): {a & b}")
print(f"Diferencia (A - B): {a - b}")
print(f"Diferencia simétrica (A ^ B): {a ^ b}")
````

````python
# Membresía (muy eficiente, O(1))
numeros_grandes = set(range(1_000_000))
print(f"999999 in numeros_grandes: {999999 in numeros_grandes}")
````

## 5. Frozenset

`frozenset` es la versión **inmutable** de un set. Se puede usar como clave de diccionario o elemento de otro set.


````python
# frozenset vs set
s = frozenset([1, 2, 3])
print(f"Frozenset: {s}")

# frozenset como clave de diccionario
permisos = {
    frozenset({"admin", "user"}): "Acceso completo",
    frozenset({"user"}): "Acceso limitado"
}
print(f"Permisos: {permisos}")
````

## 6. Estructuras complejas

### Listas de diccionarios


````python
# Lista de diccionarios (estructura común para datos tabulares)
estudiantes = [
    {"nombre": "Ana", "edad": 25, "promedio": 9.2},
    {"nombre": "Luis", "edad": 30, "promedio": 7.8},
    {"nombre": "María", "edad": 22, "promedio": 8.9},
    {"nombre": "Pedro", "edad": 28, "promedio": 6.5}
]

# Encontrar el estudiante con mejor promedio
mejor = max(estudiantes, key=lambda e: e["promedio"])
print(f"Mejor estudiante: {mejor['nombre']} con {mejor['promedio']}")
````

````python
# Diccionarios anidados
empresa = {
    "nombre": "TechCorp",
    "departamentos": {
        "ingenieria": {
            "empleados": 50,
            "proyectos": ["Alpha", "Beta", "Gamma"]
        },
        "ventas": {
            "empleados": 20,
            "proyectos": ["Delta", "Epsilon"]
        }
    }
}

print(f"Proyectos de ingeniería: {empresa['departamentos']['ingenieria']['proyectos']}")
````

````python
# Contar frecuencias con diccionarios
texto = "python es un lenguaje de programación"
palabras = texto.split()
frecuencia = {}

for palabra in palabras:
    frecuencia[palabra] = frecuencia.get(palabra, 0) + 1

print(f"Frecuencias: {frecuencia}")
````

````python
# Patrón más elegante con setdefault
frecuencia2 = {}
for palabra in palabras:
    frecuencia2.setdefault(palabra, 0)
    frecuencia2[palabra] += 1

print(f"Frecuencias: {frecuencia2}")
````

## 7. Ejercicio Práctico: Directorio de Contactos

**Contexto:** Vas a crear un directorio de contactos para un proyecto grupal universitario. La estructura será un diccionario donde la clave es la matrícula y el valor es otro diccionario con los datos del estudiante.

**Instrucciones:**

Crea un archivo `directorio.py` con las siguientes funciones:

1. `agregar_contacto(directorio, matricula, datos)` — Agrega un contacto al directorio.
2. `buscar_contacto(directorio, matricula)` — Busca y retorna los datos de un contacto.
3. `eliminar_contacto(directorio, matricula)` — Elimina un contacto.
4. `listar_contactos(directorio)` — Lista todos los contactos.
5. `contar_por_rol(directorio, rol)` — Cuenta cuántos contactos tienen un rol específico.

**Estructura de cada contacto:**
```python
{
    "nombre": "Ana",
    "email": "ana@uni.edu",
    "rol": "lider"  # lider, desarrollador, diseñador, tester
}
```

**Pistas:**
- Usa `.get()` para búsquedas seguras.
- Usa `dict.values()` para iterar sobre todos los contactos.
- Usa un contador para contar por rol.

**Restricción del ejercicio:**
- No uses clases todavía.
- Todo debe manejarse con diccionarios anidados.


````python
"""
Solución al Ejercicio Práctico
Este código debe ir en un archivo directorio.py
"""

def agregar_contacto(directorio, matricula, datos):
    """Agrega un contacto al directorio."""
    directorio[matricula] = datos
    print(f"✓ Agregado: {datos['nombre']} ({matricula})")

def buscar_contacto(directorio, matricula):
    """Busca un contacto por matrícula."""
    if matricula in directorio:
        return directorio[matricula]
    return None

def eliminar_contacto(directorio, matricula):
    """Elimina un contacto del directorio."""
    if matricula in directorio:
        nombre = directorio[matricula]['nombre']
        del directorio[matricula]
        print(f"✓ Eliminado: {nombre}")
        return True
    print(f"✗ No se encontró: {matricula}")
    return False

def listar_contactos(directorio):
    """Lista todos los contactos del directorio."""
    if not directorio:
        print("El directorio está vacío.")
        return
    print(f"\nContactos ({len(directorio)}):")
    for matricula, datos in directorio.items():
        print(f"  {matricula}: {datos['nombre']} <{datos['email']}> - {datos['rol']}")

def contar_por_rol(directorio, rol):
    """Cuenta cuántos contactos tienen un rol específico."""
    contador = 0
    for datos in directorio.values():
        if datos.get('rol') == rol:
            contador += 1
    return contador
````

````python
# Probar el directorio
directorio = {}

# Agregar contactos
agregar_contacto(directorio, "A001", {"nombre": "Ana", "email": "ana@uni.edu", "rol": "lider"})
agregar_contacto(directorio, "A002", {"nombre": "Luis", "email": "luis@uni.edu", "rol": "desarrollador"})
agregar_contacto(directorio, "A003", {"nombre": "María", "email": "maria@uni.edu", "rol": "diseñador"})
agregar_contacto(directorio, "A004", {"nombre": "Pedro", "email": "pedro@uni.edu", "rol": "desarrollador"})

# Listar
listar_contactos(directorio)
````

````python
# Buscar
print(f"Buscar A002: {buscar_contacto(directorio, 'A002')}")
print(f"Buscar A999: {buscar_contacto(directorio, 'A999')}")

# Contar por rol
print(f"\nDesarrolladores: {contar_por_rol(directorio, 'desarrollador')}")
print(f"Líderes: {contar_por_rol(directorio, 'lider')}")

# Eliminar
eliminar_contacto(directorio, "A003")
listar_contactos(directorio)
````
