## 1. Creación de strings

Los strings en Python se pueden crear con comillas simples, dobles o triples.


```python
# Comillas simples
s1 = 'Hola mundo'
print(s1)

# Comillas dobles
s2 = "Hola mundo"
print(s2)

# Strings multilínea con triple comilla
s3 = """Este es un string
que ocupa varias
líneas."""
print(s3)
```

```python
# Cuándo usar comillas simples vs dobles
# Si el texto contiene comillas simples, usa dobles
texto1 = "Don't worry, be happy"
# Si el texto contiene comillas dobles, usa simples
texto2 = 'Ella dijo "Hola"'

print(texto1)
print(texto2)
```

```python
# Raw strings (no procesan secuencias de escape)
ruta_windows = r"C:\Users\Juan\Documents"
print(ruta_windows)

# Sin raw string, necesitarías escapar
ruta_windows_escaped = "C:\\Users\\Juan\\Documents"
print(ruta_windows_escaped)
```

## 2. Secuencias de escape


```python
# Secuencias de escape comunes
print("Línea 1\nLínea 2")           # \n: nueva línea
print("Columna1\tColumna2")          # \t: tabulación
print("Dice: \"Hola\"")               # \": comilla doble
print("It\'s Python")                  # \': comilla simple
print("Ruta: C:\\Users\\test")         # \\: barra invertida
```

## 3. Inmutabilidad

Los strings son **inmutables**: no puedes cambiar un carácter individual después de crear el string.


```python
# Intentar modificar un carácter genera error
nombre = "Python"
# nombre[0] = "J"  # TypeError: 'str' object does not support item assignment

# Solución: crear un nuevo string
nuevo_nombre = "J" + nombre[1:]
print(f"Original: {nombre}")
print(f"Nuevo: {nuevo_nombre}")
```

## 4. Indexado y slicing


```python
texto = "Python es genial"
print(f"Texto: '{texto}'")
print(f"Longitud: {len(texto)}")

# Indexado (positivo)
print(f"\nPrimer carácter: '{texto[0]}'")
print(f"Tercer carácter: '{texto[2]}'")

# Indexado (negativo)
print(f"Último carácter: '{texto[-1]}'")
print(f"Penúltimo: '{texto[-2]}'")
```

```python
# Slicing [inicio:fin:paso]
print(f"texto[0:6] = '{texto[0:6]}'")      # "Python"
print(f"texto[7:] = '{texto[7:]}'")         # "es genial"
print(f"texto[:6] = '{texto[:6]}'")         # "Python"
print(f"texto[::2] = '{texto[::2]}'")       # Caracteres pares
print(f"texto[::-1] = '{texto[::-1]}'")     # Invertido
```

## 5. Operadores con strings


```python
# Concatenación con +
saludo = "Hola" + " " + "mundo"
print(saludo)

# Repetición con *
linea = "=" * 30
print(linea)

# Membresía con in / not in
texto = "Python es genial"
print(f"'Python' in texto: {'Python' in texto}")
print(f"'Java' not in texto: {'Java' not in texto}")
```

## 6. Métodos de strings


```python
# Mayúsculas y minúsculas
texto = "Hola Mundo"
print(f"upper(): '{texto.upper()}'")
print(f"lower(): '{texto.lower()}'")
print(f"capitalize(): '{texto.capitalize()}'")
print(f"title(): '{texto.title()}'")
print(f"swapcase(): '{texto.swapcase()}'")
```

```python
# Eliminar espacios
texto = "   Hola Mundo   "
print(f"strip(): '{texto.strip()}'")
print(f"lstrip(): '{texto.lstrip()}'")
print(f"rstrip(): '{texto.rstrip()}'")
```

```python
# Buscar y reemplazar
texto = "Python es genial, Python es poderoso"
print(f"find('genial'): {texto.find('genial')}")
print(f"find('Java'): {texto.find('Java')}")  # -1 si no encuentra
print(f"index('genial'): {texto.index('genial')}")
print(f"count('Python'): {texto.count('Python')}")
print(f"replace('Python', 'JavaScript'): '{texto.replace('Python', 'JavaScript')}'")
```

```python
# startswith() y endswith()
archivo = "datos.csv"
print(f"endswith('.csv'): {archivo.endswith('.csv')}")
print(f"startswith('datos'): {archivo.startswith('datos')}")
```

```python
# split() y join()
frase = "Python,Java,C++,JavaScript"
lenguajes = frase.split(",")
print(f"split: {lenguajes}")

# join: el inverso de split
reunido = " | ".join(lenguajes)
print(f"join: '{reunido}'")
```

```python
# isdigit(), isalpha(), isalnum(), isspace()
print(f"'123'.isdigit(): {'123'.isdigit()}")
print(f"'123'.isalpha(): {'123'.isalpha()}")
print(f"'abc123'.isalnum(): {'abc123'.isalnum()}")
print(f"'   '.isspace(): {'   '.isspace()}")
print(f"'Hola'.isalpha(): {'Hola'.isalpha()}")
```

## 7. Formato de strings

### f-strings (Python 3.6+, recomendado)


```python
nombre = "Ana"
edad = 25
pi = 3.14159

# Interpolación básica
print(f"Me llamo {nombre} y tengo {edad} años")

# Expresiones dentro de f-strings
print(f"El año que viene tendré {edad + 1} años")
print(f"Pi redondeado: {pi:.2f}")
```

```python
# Alineación y padding
print(f"|{'Izquierda':<15}|{'Centro':^15}|{'Derecha':>15}|")
print(f"|{'-'*15}|{'-'*15}|{'-'*15}|")
print(f"|{'texto':<15}|{'texto':^15}|{'texto':>15}|")
```

```python
# Debug con f-strings (Python 3.8+)
x = 10
y = 20
print(f"{x=}")  # Imprime "x=10"
print(f"{y=}")  # Imprime "y=20"
print(f"{x + y=}")  # Imprime "x + y=30"
```

```python
# str.format() (alternativa)
print("Me llamo {} y tengo {} años".format(nombre, edad))
print("Me llamo {n} y tengo {e} años".format(n=nombre, e=edad))
print("Pi: {:.2f}".format(pi))
```

## 8. `ord()` y `chr()`


```python
# ord(): carácter a código Unicode
print(f"ord('A') = {ord('A')}")
print(f"ord('a') = {ord('a')}")
print(f"ord('0') = {ord('0')}")
print(f"ord('ñ') = {ord('ñ')}")
print(f"ord('水') = {ord('水')}")  # Chino

# chr(): código a carácter
print(f"\nchr(65) = '{chr(65)}'")
print(f"chr(97) = '{chr(97)}'")
print(f"chr(241) = '{chr(241)}'")  # ñ
```

## 9. Ejercicio Práctico: Procesador de Texto

**Contexto:** Vas a crear un script que normalice y valide datos de entrada (correos, nombres, números de teléfono) usando solo métodos de strings.

**Instrucciones:**

Crea un archivo `procesador_texto.py` con las siguientes funciones:

1. `normalizar_nombre(nombre)` — Elimina espacios extra y capitaliza cada palabra.
2. `validar_email(email)` — Verifica formato básico: contiene exactamente un `@` y al menos un `.` después del `@`.
3. `normalizar_telefono(telefono)` — Elimina caracteres no numéricos y agrega prefijo `+52` si no lo tiene.
4. `limpiar_texto(texto)` — Elimina espacios múltiples y caracteres especiales (excepto letras, números y espacios).

**Pistas:**
- Usa `.split()` y `.join()` para eliminar espacios múltiples.
- Usa `.strip()` para eliminar espacios al inicio y final.
- Para el email, usa `.count('@')` y `.find('@')`.
- Para el teléfono, usa un bucle o `.replace()` para eliminar caracteres.

**Restricción del ejercicio:**
- No uses expresiones regulares (`re`).
- Solo métodos estándar de strings.


```python
"""
Solución al Ejercicio Práctico
Este código debe ir en un archivo procesador_texto.py
"""

def normalizar_nombre(nombre):
    """Normaliza un nombre: sin espacios extra y capitalizado."""
    # Eliminar espacios al inicio/final y múltiples espacios
    partes = nombre.split()
    return ' '.join(parte.capitalize() for parte in partes)

def validar_email(email):
    """Valida formato básico de email."""
    if email.count('@') != 1:
        return False
    pos_arroba = email.find('@')
    # Debe haber al menos un punto después del @
    if '.' not in email[pos_arroba:]:
        return False
    # No debe empezar ni terminar con @
    if pos_arroba == 0 or pos_arroba == len(email) - 1:
        return False
    return True

def normalizar_telefono(telefono):
    """Normaliza un número de teléfono."""
    # Eliminar todo lo que no sea dígito
    digitos = ''.join(c for c in telefono if c.isdigit())
    if not digitos:
        return None
    # Agregar prefijo +52 si no lo tiene
    if not digitos.startswith('52'):
        digitos = '52' + digitos
    return '+' + digitos

def limpiar_texto(texto):
    """Elimina caracteres especiales y espacios múltiples."""
    # Solo letras, números y espacios
    limpio = ''.join(c for c in texto if c.isalnum() or c.isspace())
    # Eliminar espacios múltiples
    return ' '.join(limpio.split())
```

```python
# Probar las funciones
print("--- Normalizar nombres ---")
nombres_test = ["  juan PÉREZ  ", "MARÍA   lópez", "ana   GARCIA"]
for nombre in nombres_test:
    print(f"  '{nombre}' → '{normalizar_nombre(nombre)}'")

print("\n--- Validar emails ---")
emails_test = ["ana@uni.edu", "invalido.com", "@sinusuario.com", "sin@dominio", "doble@@arroba.com"]
for email in emails_test:
    valido = "✓" if validar_email(email) else "✗"
    print(f"  {valido} {email}")

print("\n--- Normalizar teléfonos ---")
tels_test = ["5512345678", "+52 55 1234-5678", "55-1234-5678", "abc123def456"]
for tel in tels_test:
    resultado = normalizar_telefono(tel)
    print(f"  '{tel}' → '{resultado}'")

print("\n--- Limpiar texto ---")
textos_test = ["Hola!! mundo??", "Python   es    genial", "Texto#con$caracteres@especiales"]
for texto in textos_test:
    print(f"  '{texto}' → '{limpiar_texto(texto)}'")
```
