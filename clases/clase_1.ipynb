## 1. Historia de Python

Python fue creado por **Guido van Rossum** en 1991 en los Países Bajos. El nombre "Python" no viene de la serpiente, sino del grupo de comedia británico **Monty Python** (del que Guido era fan).

### La influencia de ABC

Python fue diseñado como sucesor del lenguaje **ABC**, un lenguaje educativo creado en los años 80 en el **CWI** (Centro de Matemáticas y Ciencias de la Computación de Ámsterdam). De ABC, Python heredó:

- **Sintaxis clara y legible**: pocos símbolos, palabras en inglés natural.
- **Indentación significativa**: el código se estructura visualmente.
- **Tipos de alto nivel incorporados**: listas, diccionarios, strings como tipos primarios.
- **Intérprete interactivo**: ideal para aprender y experimentar.

> **Dato curioso:** Guido van Rossum fue conocido como el "Benevolente Dictador Vitalicio" (BDFL) de Python hasta 2018, cuando se retiró de ese rol. Hoy el lenguaje lo guía un consejo de cinco miembros.

### Línea temporal de Python

| Año | Evento |
|---|---|
| 1980s | Guido trabaja en ABC en el CWI |
| 1989 | Inicia el desarrollo de Python durante las vacaciones de Navidad |
| 1991 | Python 0.9.0 publicado en alt.sources |
| 2000 | Python 2.0 publicado |
| 2008 | Python 3.0 publicado (no retrocompatible) |
| 2020 | Python 2 llega a su fin de vida |
| 2024+ | Python 3.12+ es la versión actual, con mejoras de velocidad y typing |

## 2. Las dos grandes ventajas de Python

### Ventaja 1: Sintaxis sencilla

Python está diseñado para que el código sea **leíble como pseudocódigo en inglés**. Compara el mismo programa en C y en Python:

**Hola Mundo en C:**
```c
#include <stdio.h>

int main() {
    printf("Hola, mundo!\n");
    return 0;
}
```

**Hola Mundo en Python:**
```python
print("Hola, mundo!")
```

La diferencia es enorme. En C necesitas:
- Incluir librerías (`#include`)
- Declarar el tipo de retorno de `main` (`int`)
- Escribir la firma completa de la función
- Usar punto y coma al final
- Usar `\n` para el salto de línea

En Python solo necesitas una línea.

**Más comparaciones:**

| Concepto | C | Python |
|---|---|---|
| Imprimir texto | `printf("Hola\n");` | `print("Hola")` |
| Declarar variable | `int x = 5;` | `x = 5` |
| Comentario | `// comentario` | `# comentario` |
| Bloque de código | `{ ... }` | `indentación` |
| Leer input | `scanf("%d", &x);` | `x = int(input())` |

### Ventaja 2: Librería estándar extensa

Python viene con **"batteries included"** (pilas incluidas): trae una enorme cantidad de módulos para hacer casi cualquier cosa sin instalar nada extra:

- `os`, `sys`: interactuar con el sistema operativo
- `math`, `random`: operaciones matemáticas
- `datetime`: manejar fechas y horas
- `json`, `csv`: leer y escribir formatos de datos
- `collections`, `itertools`: estructuras de datos avanzadas
- `pathlib`: manipulación de rutas de archivos
- `re`: expresiones regulares
- `urllib`, `http`: peticiones web
- `sqlite3`: base de datos SQLite
- `unittest`, `pytest`: testing
- ... ¡y muchos más!

> **Como consecuencia de la sintaxis sencilla:** Python se ha convertido en el lenguaje dominante en ciencia de datos, machine learning, automatización, scripting y desarrollo web, lo que ha generado un ecosistema de **paquetes externos** gigantesco (más de 400,000 en PyPI).

## 3. Las dos grandes desventajas de Python

### Desventaja 1: Velocidad de ejecución

Python es un lenguaje **interpretado** (el código se ejecuta línea por línea) y **dinámicamente tipado** (los tipos se verifican en tiempo de ejecución). Esto lo hace más lento que lenguajes compilados como C o C++.

**Comparación aproximada de velocidad** (operaciones por segundo, normalizado):

| Lenguaje | Velocidad relativa | Compilado/Interpretado |
|---|---|---|
| C | 100x | Compilado |
| C++ | 95x | Compilado |
| Java | 50x | Compilado a bytecode |
| C# | 40x | Compilado a bytecode |
| **Python** | **5-10x** | **Interpretado** |
| JavaScript | 15-20x | Interpretado (con JIT) |

> **Pero no todo es velocidad:** Para la mayoría de tareas (desarrollo web, análisis de datos, scripting, IA), la velocidad de Python es **más que suficiente**. Además, cuando necesitas velocidad, puedes escribir las partes críticas en C y llamarlas desde Python.

### Desventaja 2: Requiere gestores de paquetes externos

Aunque Python viene con una librería estándar rica, para proyectos reales necesitas instalar paquetes de terceros. Esto requiere un **gestor de paquetes** (como `pip` o `uv`) y un concepto llamado **entornos virtuales** para aislar dependencias entre proyectos.

En C, normalmente enlazas contra librerías del sistema (`-lm` para math, `-lpthread` para threads). En Python, el flujo es:

```bash
# Instalar un paquete
uv add requests

# Activar un entorno virtual
source .venv/bin/activate  # Linux/macOS
.venv\Scripts\activate     # Windows
```

> **La buena noticia:** Los gestores modernos como **uv** (que veremos hoy) son increíblemente rápidos y hacen este proceso casi indoloro.

## 4. Alto nivel vs bajo nivel

Los lenguajes de programación se pueden clasificar por su **nivel de abstracción** respecto al hardware:

| Nivel | Características | Ejemplos |
|---|---|---|
| **Bajo nivel** | Cercano al hardware, acceso directo a memoria, instrucciones simples | Ensamblador, C |
| **Nivel medio** | Mezcla de abstracción y control, permite manipular memoria | C++, C#, Java |
| **Alto nivel** | Abstracción del hardware, gestión automática de memoria, sintaxis natural | Python, Ruby, JavaScript |
| **Muy alto nivel** | Abstracción extrema, código declarativo | SQL, HTML, Prolog |

### ¿Dónde está Python?

Python es un lenguaje de **alto nivel**. Esto significa que:

1. **Gestión automática de memoria:** No necesitas `malloc()` ni `free()`. El **recolector de basura** (garbage collector) se encarga de liberar memoria cuando los objetos ya no se usan.

2. **Tipos de datos de alto nivel:** Listas, diccionarios y conjuntos son tipos primarios del lenguaje, no estructuras que tengas que construir manualmente con arrays y punteros.

3. **Interpretado:** No necesitas compilar a código máquina. El intérprete lee y ejecuta tu código directamente.

4. **Multiparadigma:** Soporta programación imperativa, orientada a objetos, funcional y procedural.

### Comparación C vs Python

| Aspecto | C | Python |
|---|---|---|
| Gestión de memoria | Manual (`malloc`/`free`) | Automática (garbage collector) |
| Compilación | Necesaria (`gcc`) | No necesaria (interpretado) |
| Tipos de datos | Primitivos (`int`, `float`, `char`) | De alto nivel (`list`, `dict`, `set`) |
| Arrays | Tamaño fijo o `realloc` | Dinámicos (listas) |
| Cadenas | `char[]` terminado en `'\0'` | Objetos `str` con Unicode |
| Velocidad | Muy rápida | Más lenta |
| Curva de aprendizaje | Empinada | Suave |

## 5. El intérprete de Python

Python es un lenguaje **interpretado**. Esto significa que hay un programa (el **intérprete de Python**) que lee tu código y lo ejecuta directamente, sin necesidad de compilarlo a código máquina previamente.

### ¿Cómo funciona?

```
Tu código (.py)  →  Intérprete de Python  →  Ejecución directa
     ↓                       ↓                       ↓
print("Hola")     Lee línea por línea    Imprime "Hola"
```

Internamente, el intérprete:
1. **Compila** tu código a **bytecode** (instrucciones de bajo nivel para una máquina virtual).
2. **Ejecuta** el bytecode en la **Python Virtual Machine (PVM)**.

Este proceso es transparente para ti: simplemente escribes y ejecutas.

### El REPL (Read-Eval-Print Loop)

Python trae un **intérprete interactivo** llamado REPL. Para usarlo:

```bash
python
```

Verás algo como:
```
Python 3.12.x (main, ...)
>>> 
```

Ahora puedes escribir código Python directamente:

```python
>>> 2 + 2
4
>>> print("Hola")
Hola
>>> nombre = "Python"
>>> nombre
'Python'
```

El REPL evalúa cada expresión inmediatamente. Es perfecto para **experimentar y aprender**.

> **Tip:** En Jupyter, cada celda de código es como un mini-REPL. Puedes ejecutar expresiones directamente.

````python
2 + 2
````

## 6. Instalación de UV y Jupyter

**UV** es un gestor de paquetes y proyectos de Python extremadamente rápido, escrito en Rust. Reemplaza a `pip`, `pip-tools`, `pipx`, `pyenv`, `virtualenv` y `poetry` con una sola herramienta.

### Instalación de UV

**Windows (PowerShell):**
```powershell
powershell -ExecutionPolicy ByPass -c "irm https://astral.sh/uv/install.ps1 | iex"
```

**macOS/Linux:**
```bash
curl -LsSf https://astral.sh/uv/install.sh | sh
```

### Crear un proyecto con entorno virtual

```bash
# Crear carpeta del proyecto
mkdir mi_proyecto
cd mi_proyecto

# Inicializar proyecto
uv init

# Crear entorno virtual
uv venv

# Instalar Jupyter
uv add jupyter ipykernel
```

### Iniciar Jupyter Notebook

```bash
uv run jupyter notebook
```

Esto abrirá Jupyter en tu navegador. Crea un nuevo notebook seleccionando **Python 3** como kernel.

> **Nota:** Si usas VS Code, solo necesitas la extensión de Jupyter. VS Code detectará automáticamente tu entorno `.venv` y podrás seleccionar el kernel desde la esquina superior derecha de cada notebook.

## 7. Tu primer script

Un script en Python es un archivo de texto con extensión `.py` que contiene código Python. Vamos a crear nuestro primer script.

### Hola Mundo en Python

Crea un archivo llamado `hola.py` con este contenido:

```python
# mi_primer_script.py
# Este es un comentario: el intérprete lo ignora

print("¡Hola, mundo!")
print("Bienvenido al curso de Python")
```

Para ejecutarlo desde la terminal:

```bash
uv run python hola.py
```

Deberías ver:
```
¡Hola, mundo!
Bienvenido al curso de Python
```

### Estructura básica de un script

Un script de Python típicamente tiene:

1. **Comentarios** (líneas que empiezan con `#`): explican el código.
2. **Imports** (al inicio): módulos que se necesitan.
3. **Definiciones** (funciones, clases): bloques reutilizables.
4. **Código principal**: la lógica que se ejecuta.

```python
# mi_script.py

# 1. Comentario que explica el propósito del script
"""Docstring: descripción del módulo."""

# 2. Imports
import os
import sys

# 3. Definiciones
def saludar(nombre):
    return f"Hola, {nombre}!"

# 4. Código principal
if __name__ == '__main__':
    nombre = input("¿Cómo te llamas? ")
    print(saludar(nombre))
```

> **Por ahora** no te preocupes por entender todo: `if __name__ == '__main__':` y las funciones (`def`) las veremos en clases posteriores. Hoy solo nos enfocamos en `print()`.

## 8. La función `print()`

`print()` es la función más básica de Python. Envía texto a la salida estándar (la consola).

### Uso básico

```python
print("Hola, mundo")
```

**Salida:**
```
Hola, mundo
```

### Características principales

| Uso | Ejemplo | Resultado |
|---|---|---|
| Un solo valor | `print("Hola")` | `Hola` |
| Varios valores (separados por espacio) | `print("Hola", "mundo")` | `Hola mundo` |
| Parámetro `sep` (separador personalizado) | `print("a", "b", "c", sep="-")` | `a-b-c` |
| Parámetro `end` (cambia el final) | `print("Hola", end=" ")` | `Hola ` (sin salto de línea) |
| Imprimir variables | `print(f"Tengo {25} años")` | `Tengo 25 años` |
| Representación con `!r` | `print(f"{'Hola'!r}")` | `'Hola'` |

> **Nota:** Las f-strings (formato con `f""`) y los conceptos más avanzados de `print()` los exploraremos en profundidad en clases posteriores. Por ahora, quédate con el uso básico.

````python
print("¡Hola, mundo!")
print("Bienvenido al curso de Python")
print("Este es tu primer script")
````

````python
print("Hola", "mundo")                    # Varios valores separados por espacio
print("a", "b", "c", sep="-")            # Separador personalizado
print("Hola", end=" ")                   # Sin salto de línea al final
print("mundo")                            # Continúa en la misma línea
print()                                   # Línea vacía
print(f"2 + 2 = {2 + 2}")                # f-string con expresión
print(f"{'Python'!r}")                   # Representación con comillas
````

## 9. Herramientas de IA como asistente de aprendizaje

La inteligencia artificial es una herramienta poderosa para aprender programación. Puedes usar herramientas como **ChatGPT**, **Claude** o **GitHub Copilot** como tutores personales.

### Usos efectivos de la IA en programación

1. **Explicar código** que no entiendes: "Explícame qué hace esta función línea por línea"
2. **Generar ejemplos** de uso: "Dame 3 ejemplos de cómo usar listas en Python"
3. **Sugerir nombres** descriptivos para variables y funciones
4. **Refactorizar** código: "¿Cómo puedo hacer esta función más legible?"
5. **Buscar errores**: "Mi código da este error, ¿qué puede estar mal?"
6. **Aprender conceptos nuevos**: "Explícame qué son los decoradores con un ejemplo simple"

### La regla de oro

> **La IA puede generar código, pero no puede evaluar si ese código es correcto, eficiente o seguro. Esa responsabilidad recae en ti.**

**Verifica siempre:**
- ¿El código hace lo que esperas?
- ¿Entiendes cada línea?
- ¿Hay casos edge que no maneja?
- ¿Es seguro?

Un buen workflow es:
1. Intenta resolver el problema tú mismo primero
2. Si te atascas, pide ayuda a la IA con una pregunta específica
3. Verifica la solución propuesta
4. Adapta el código a tu estilo y necesidad

> **En este curso**, te animo a usar IA como asistente, pero **siempre entendiendo el código** que produce. Si copias y pegas sin entender, no estás aprendiendo.
