## 1. ¿Qué es la POO?

La **Programación Orientada a Objetos** organiza el código alrededor de **objetos** que combinan:
- **Datos** (atributos)
- **Comportamiento** (métodos)

En lugar de tener variables y funciones sueltas, agrupamos lo que está relacionado en **clases** (plantillas) que producen **objetos** (instancias).

### Analogía con C

Una clase es como un `struct` en C, pero que también contiene las funciones que operan sobre esos datos:

```c
// En C
struct Estudiante {
    char nombre[50];
    int edad;
    float promedio;
};
void mostrar_estudiante(struct Estudiante e) { ... }
```

En Python, datos Y funciones viven juntos en la clase.

## 3. El método `__init__` y `self`

El método `__init__` es el **constructor**: se llama automáticamente cuando creas un objeto. El parámetro `self` es una referencia al objeto que se está creando.


````python
class Estudiante:
    """Representa a un estudiante del curso."""
    # Espacio para declarar atributos de clase (si los hubiera)
    # Ejemplo: contador_estudiantes = 0

    def __init__(self, nombre, edad, matricula):
        """Inicializa un nuevo estudiante."""
        # Atributos
        self.nombre = nombre
        self.edad = edad
        self.matricula = matricula

    # No existe en python control de acceso pero puedes usar convenciones
    def _atributo_privado(self):
        """Método privado (convención)."""
        return "Este es un método privado"
    
    # Métodos
    def presentarse(self):
        """El estudiante se presenta."""
        return f"Soy {self.nombre}, tengo {self.edad} años, matrícula {self.matricula}"
    
    # Dunder method de impresión
    def __str__(self):
        return f"Estudiante: {self.nombre}, Edad: {self.edad}, Matrícula: {self.matricula}"

# Crear objetos
ana = Estudiante("Ana", 25, "A001")
luis = Estudiante("Luis", 30, "B002")

print(ana.presentarse())
print(luis.presentarse())
print(ana)
print(luis)
````

````python
# Acceder a los atributos
print(f"Nombre: {ana.nombre}")
print(f"Edad: {ana.edad}")
print(f"Matrícula: {ana.matricula}")
````

````python
# Los atributos se pueden modificar directamente (a menos que los hagamos privados)
ana.edad = 26
print(f"\nNueva edad de Ana: {ana.edad}")
````

## 4. Atributos de instancia y de clase

- **Atributos de instancia**: únicos para cada objeto (como `nombre` arriba).
- **Atributos de clase**: compartidos por todas las instancias.

## 5. Métodos: instancia, estáticos, de clase


````python
class Calculadora:
    """Calculadora con diferentes tipos de métodos."""
    
    pi = 3.14159  # Atributo de clase
    
    def __init__(self, nombre):
        self.nombre = nombre  # Atributo de instancia
    
    def sumar(self, a, b):  # Método de instancia
        """Método de instancia: usa self."""
        return a + b
    
    # staticmethod y classmethod no usan self, pero classmethod usa cls
    @staticmethod
    def multiplicar(a, b):  # Método estático
        """Método estático: no usa self ni cls."""
        return a * b
    
    @classmethod
    def con_nombre(cls, sufijo):  # Método de clase
        """Método de clase: recibe la clase como cls."""
        return cls(f"Calculadora-{sufijo}")

# Uso
calc = Calculadora("MiCalc")
print(f"sumar(3, 5): {calc.sumar(3, 5)}")
print(f"multiplicar(3, 5): {Calculadora.multiplicar(3, 5)}")  # Sin instancia
calc2 = Calculadora.con_nombre("v2")
print(f"Nombre: {calc2.nombre}")
````

## 6. Propiedades con `@property`

Las propiedades permiten controlar el acceso a atributos con getters y setters.


````python
class Temperatura:
    """Temperatura con validación en el setter."""
    
    def __init__(self, celsius=0):
        self._celsius = celsius  # Atributo "privado"

    def celsius_funcion(self):
        """Getter: obtener temperatura en Celsius."""
        return self._celsius
    
    @property
    def celsius(self):
        """Getter: obtener temperatura en Celsius."""
        return f"{self._celsius}°C"
    
    @celsius.setter
    def celsius(self, valor):
        """Setter: validar antes de asignar."""
        if valor < -273.15:
            raise ValueError("Temperatura por debajo del cero absoluto")
        self._celsius = valor
    
    @property
    def fahrenheit(self):
        """Calcular Fahrenheit (solo lectura)."""
        return self._celsius * 9 / 5 + 32

# Uso
temp = Temperatura(25)
# Utilizamos los properties para acceder a los atributos sin llamar a métodos directamente
print(f"Celsius_funcion: {temp.celsius_funcion()}")
print(f"Celsius: {temp.celsius}")
print(f"Fahrenheit: {temp.fahrenheit}")

temp.celsius = 30
print(f"\nNuevo Celsius: {temp.celsius}")
print(f"Nuevo Fahrenheit: {temp.fahrenheit}")

try:
    temp.celsius = -300
except ValueError as e:
    print(f"\nError: {e}")
````

## 7. Encapsulamiento: público, protegido, privado

Python no tiene privacidad real, pero usa convenciones:

- `nombre` → público (cualquiera accede)
- `_nombre` → "protegido" (convención: no usar fuera de la clase)
- `__nombre` → "privado" (name mangling: `_Clase__nombre`)


````python
class CuentaBancaria:
    """Cuenta con encapsulamiento."""
    
    def __init__(self, titular, saldo_inicial=0):
        self.titular = titular            # público
        self._saldo = saldo_inicial       # "protegido" (convención)
        self.__secreto = "1234"          # "privado" (name mangling)
    
    def depositar(self, monto):
        if monto > 0:
            self._saldo += monto
    
    def obtener_saldo(self):
        return self._saldo

# Uso
cuenta = CuentaBancaria("Ana", 1000)
print(f"Titular: {cuenta.titular}")           # público: OK
print(f"Saldo: {cuenta.obtener_saldo()}")      # usar getter
# print(cuenta.__secreto)                    # AttributeError
print(f"Acceso 'privado': {cuenta._CuentaBancaria__secreto}")  # name mangling
````

## 8. Composición y agregación

**Composición** es cuando una clase contiene instancias de otra clase. Es la relación "tiene un".

**Agregación** es cuando una clase recibe instancias de otra clase. Es la relación "tiene un", pero no depende de la otra para existir.

````python
class Motor:
    def __init__(self, potencia):
        self.potencia = potencia
    
    def encender(self):
        return f"Motor de {self.potencia}HP encendido"

class Coche:
    """Coche COMPUESTO por un motor (composición)."""
    
    def __init__(self, marca, potencia_motor):
        self.marca = marca
        # Composición: el coche TIENE UN motor
        self.motor = Motor(potencia_motor)
    
    def arrancar(self):
        return f"{self.marca}: {self.motor.encender()}"

# Uso
coche = Coche("Toyota", 150)
print(coche.arrancar())



````

````python
class Auto:
    def __init__(self, marca):
        self.marca = marca

    def encender(self):
        return "El auto está encendido"


class Celular:
    def __init__(self, modelo):
        self.modelo = modelo

    def llamar(self):
        return "Realizando llamada"


class Persona:
    """ Persona que tiene un auto y un celular (agregación)."""
    
    def __init__(self, nombre, auto, celular):
        self.nombre = nombre
        # Los objetos se pasan desde afuera
        self.auto = auto
        self.celular = celular

    def usar_auto(self):
        if self.auto is None:
            return "No tengo auto"
        return self.auto.encender()

    def usar_celular(self):
        if self.celular is None:
            return "No tengo celular"
        return self.celular.llamar()
    
# Uso
auto = Auto("Toyota")
celular = Celular("iPhone")
persona = Persona("Ana", auto, celular)
print(persona.usar_auto())
print(persona.usar_celular())
# Persona no tiene auto

persona_2 = Persona("Luis", None, celular)
print(persona_2.usar_auto())  # Esto generará un error porque auto es None, se puede manejar con validación si se desea.
````

## 9. Dataclasses

Las **dataclasses** simplifican la creación de clases que principalmente almacenan datos.


````python
from dataclasses import dataclass, field

@dataclass
class Punto:
    """Punto en 2D con dataclass."""
    x: float
    y: float
    nombre: str = "Origen"  # Valor por defecto

# Crear instancias
p1 = Punto(3, 4)
p2 = Punto(0, 0, "Centro")
p3 = Punto(3, 4)  # Igual que p1

print(p1)
print(p2)
print(f"\nIgualdad: {p1 == p3}")  # __eq__ automático
````

````python
# Dataclass con más funcionalidad
@dataclass
class Producto:
    nombre: str
    precio: float
    stock: int = 0
    tags: list = field(default_factory=list)  # Para listas vacías

# Crear y modificar
prod = Producto("Laptop", 1500, 5)
prod.tags.append("electrónica")
print(prod)
````

## 10. Métodos mágicos básicos

Los **métodos mágicos** (o dunder) son métodos con doble guión bajo que Python llama automáticamente.


````python
class Vector:
    """Vector 2D con métodos mágicos."""
    
    def __init__(self, x, y):
        self.x = x
        self.y = y
    
    # Dunder method para representación
    def __repr__(self):
        return f"Vector({self.x}, {self.y})"
    
    # Dunder method para impresión
    def __str__(self):
        return f"({self.x}, {self.y})"
    
    # Dunder method para sobrecargar el operador ==
    def __eq__(self, otro):
        return self.x == otro.x and self.y == otro.y
    
    # Dunder method para sobrecargar el operador +
    def __add__(self, otro):
        return Vector(self.x + otro.x, self.y + otro.y)
    
    # Dunder method para sobrecargar el operador len
    def __len__(self):
        return 2  # Un vector 2D tiene "2 dimensiones"

# Uso
v1 = Vector(3, 4)
v2 = Vector(1, 2)
v3 = v1 + v2

print(f"repr: {repr(v1)}")
print(f"str: {v1}")
print(f"v1 + v2 = {v3}")
print(f"v1 == Vector(3, 4): {v1 == Vector(3, 4)}")
print(f"len(v1) = {len(v1)}")
````

## 11. Ejercicio Práctico: Gestor de Estudiantes con POO

**Contexto:** Vas a crear una clase `Estudiante` que encapsule toda la lógica de gestión de calificaciones que antes hacíamos con diccionarios.

**Instrucciones:**

Crea un archivo `estudiante.py` con una clase `Estudiante` que tenga:

1. Atributos: `nombre`, `matricula`, `calificaciones` (diccionario materia→nota).
2. Método `agregar_calificacion(materia, nota)` con validación (0-10).
3. Método `obtener_promedio()` que retorna el promedio de todas las calificaciones.
4. Método `obtener_materias_aprobadas()` que retorna una lista de las materias con nota >= 7.
5. Método `__str__` que muestre nombre, matrícula y promedio.
6. Método `__repr__` que muestre la representación para debugging.

**Pistas:**
- Usa `@property` para el promedio si quieres que se calcule automáticamente.
- En `agregar_calificacion`, valida que la nota esté entre 0 y 10.
- Si la materia ya existe, actualiza la nota; si no, agrégala.

**Restricción del ejercicio:**
- No uses herencia todavía.
- No uses dataclasses (usa una clase regular para practicar).


````python
"""
Solución al Ejercicio Práctico
Este código debe ir en un archivo estudiante.py
"""

class Estudiante:
    """Representa a un estudiante con sus calificaciones."""
    
    def __init__(self, nombre, matricula):
        self.nombre = nombre
        self.matricula = matricula
        self.calificaciones = {}  # {materia: nota}
    
    def agregar_calificacion(self, materia, nota):
        """Agrega o actualiza una calificación."""
        if not 0 <= nota <= 10:
            raise ValueError(f"Nota debe estar entre 0 y 10, recibido: {nota}")
        self.calificaciones[materia] = nota
    
    def obtener_promedio(self):
        """Calcula el promedio de todas las calificaciones."""
        if not self.calificaciones:
            return 0.0
        return sum(self.calificaciones.values()) / len(self.calificaciones)
    
    def obtener_materias_aprobadas(self):
        """Retorna lista de materias con nota >= 7."""
        return [materia for materia, nota in self.calificaciones.items() if nota >= 7]
    
    def __str__(self):
        return f"Estudiante: {self.nombre} (Mat: {self.matricula}) | Promedio: {self.obtener_promedio():.2f}"
    
    def __repr__(self):
        return f"Estudiante(nombre='{self.nombre}', matricula='{self.matricula}')"
````

````python
# Probar la clase Estudiante
ana = Estudiante("Ana López", "A001")
ana.agregar_calificacion("Cálculo", 9.0)
ana.agregar_calificacion("Programación", 8.5)
ana.agregar_calificacion("Física", 6.5)
ana.agregar_calificacion("Química", 9.5)

print(str(ana))
print(f"repr: {repr(ana)}")
print(f"\nPromedio: {ana.obtener_promedio():.2f}")
print(f"Materias aprobadas: {ana.obtener_materias_aprobadas()}")

# Validación
try:
    ana.agregar_calificacion("Biología", 15)
except ValueError as e:
    print(f"\nError: {e}")
````
