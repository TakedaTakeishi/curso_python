### List comprehensions con múltiples `for`


```python
# Aplanar una matriz
matriz = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
plano = [numero for fila in matriz for numero in fila]
print(f"Matriz aplanada: {plano}")

# Producto cartesiano
colores = ['rojo', 'verde']
tallas = ['S', 'M', 'L']
combinaciones = [f"{c}-{t}" for c in colores for t in tallas]
print(f"Combinaciones: {combinaciones}")
```

```python
# Dict comprehension con condición
numeros = range(1, 11)
cuadrados = {n: n**2 for n in numeros if n % 2 == 0}
print(f"Cuadrados de pares: {cuadrados}")
```

```python
# Nested comprehensions: transponer una matriz
matriz = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
transpuesta = [[fila[i] for fila in matriz] for i in range(len(matriz[0]))]
print("Matriz original:")
for fila in matriz:
    print(fila)
print("\nTranspuesta:")
for fila in transpuesta:
    print(fila)
```

## 2. Expresiones regulares (regex)

Las **regex** son patrones para buscar y manipular texto. Python tiene el módulo `re`.


```python
import re

# Buscar un patrón
texto = "Mi email es ana@uni.edu y mi teléfono es 55-1234-5678"
emails = re.findall(r'[\w.-]+@[\w.-]+', texto)
print(f"Emails encontrados: {emails}")

# Buscar y reemplazar
texto_limpio = re.sub(r'\d{2}-\d{4}-\d{4}', '[TELÉFONO]', texto)
print(f"Texto limpio: {texto_limpio}")
```

```python
# Patrones comunes
patrones = {
    'email': r'^[\w.-]+@[\w.-]+\.\w+$',
    'teléfono': r'^\+?[\d\s-]{10,}$',
    'URL': r'^https?://[\w.-]+$',
    'fecha': r'^\d{2}/\d{2}/\d{4}$',
}

# Probar
test_strings = [
    "ana@uni.edu",
    "+52 55 1234-5678",
    "https://python.org",
    "15/03/2026",
    "no es válido"
]

for string in test_strings:
    for tipo, patron in patrones.items():
        if re.match(patron, string):
            print(f"  '{string}' es {tipo}")
            break
    else:
        print(f"  '{string}' no coincide con ningún patrón")
```

```python
# Grupos de captura
fecha_str = "Hoy es 15/03/2026 y mañana será 16/03/2026"
patron = r'(\d{2})/(\d{2})/(\d{4})'
fechas = re.findall(patron, fecha_str)
print(f"Fechas encontradas: {fechas}")
for dia, mes, año in fechas:
    print(f"  {dia}/{mes}/{año}")
```

## 3. Programación asíncrona: `async`/`await`

La **asincronía** permite ejecutar operaciones de I/O de forma concurrente sin bloquear el programa.


```python
import asyncio
import time

async def tarea(nombre, duracion):
    """Simula una tarea asíncrona."""
    print(f"  [{nombre}] Iniciando...")
    await asyncio.sleep(duracion)  # Espera sin bloquear
    print(f"  [{nombre}] Completada después de {duracion}s")
    return f"Resultado de {nombre}"

# Ejecutar tareas en paralelo
async def main():
    print("Iniciando 3 tareas en paralelo:\n")
    inicio = time.time()
    
    # gather ejecuta las 3 tareas concurrentemente
    resultados = await asyncio.gather(
        tarea("Tarea 1", 2),
        tarea("Tarea 2", 1),
        tarea("Tarea 3", 3)
    )
    
    fin = time.time()
    print(f"\nTodas completadas en {fin - inicio:.2f} segundos")
    print(f"Resultados: {resultados}")

# Ejecutar
await main()
```

### Corrutinas vs funciones normales


```python
# async define una corrutina
async def corrutina():
    return "Soy asíncrona"

# Llamarla no la ejecuta, solo la crea
co = corrutina()
print(f"Tipo: {type(co).__name__}")  # coroutine

# Para ejecutarla, usar asyncio.run() o await
resultado = await co
print(f"Resultado: {resultado}")
```

## 4. Principios de Clean Code

**Clean Code** son principios para escribir código legible, mantenible y profesional.


```python
# ❌ Mal código: nombres crípticos, sin estructura
def f(x, y):
    a = x * 0.1
    b = y - a
    if b < 0:
        b = 0
    return x + b

# ✓ Buen código: nombres descriptivos, intención clara
def calcular_precio_con_descuento(precio_base, porcentaje_descuento):
    """Aplica un descuento al precio base."""
    descuento = precio_base * (porcentaje_descuento / 100)
    precio_final = precio_base - descuento
    return max(precio_final, 0)  # No puede ser negativo
```

```python
# Principios de Clean Code en Python

# 1. Nombres descriptivos (no usar abreviaturas)
# ❌
def calc_iva(amt, rate):
    return amt * rate

# ✓
def calcular_iva(monto, tasa):
    """Calcula el IVA sobre un monto."""
    return monto * tasa
```

## 7. Ejercicio Integrador: Scraper Inteligente

**Contexto:** Vas a crear un script que use **regex** para extraer información de un bloque de texto, **list comprehensions** para procesar los datos, y **type hints** para documentar el código.

**Instrucciones:**

Crea un archivo `scraper.py` con:

1. `extraer_emails(texto: str) -> list[str]` — Extrae todos los emails de un texto.
2. `extraer_telefonos(texto: str) -> list[str]` — Extrae todos los teléfonos (formatos variados).
3. `extraer_urls(texto: str) -> list[str]` — Extrae todas las URLs (http/https).
4. `procesar_texto(texto: str) -> dict` — Combina todo en un diccionario.
5. `main()` — Punto de entrada que lee un texto de ejemplo e imprime los resultados.

**Pistas:**
- Usa `re.findall()` con patrones apropiados.
- Usa `set()` para eliminar duplicados.
- Usa list comprehensions para limpiar los datos.
- Usa type hints en todas las funciones.

**Restricción del ejercicio:**
- No uses librerías externas como `beautifulsoup`.
- Solo el módulo `re` de la biblioteca estándar.
- Usa buenas prácticas de Clean Code.


```python
"""
Solución al Ejercicio Integrador
Este código debe ir en un archivo scraper.py
"""

import re
from typing import List, Dict

def extraer_emails(texto: str) -> List[str]:
    """Extrae todos los emails únicos del texto."""
    patron = r'[\w.-]+@[\w.-]+\.\w+'
    return list(set(re.findall(patron, texto)))

def extraer_telefonos(texto: str) -> List[str]:
    """Extrae todos los teléfonos únicos del texto."""
    patron = r'\+?\d{1,3}[-.\s]?\(?\d{1,4}\)?[-.\s]?\d{3,4}[-.\s]?\d{3,4}'
    return list(set(re.findall(patron, texto)))

def extraer_urls(texto: str) -> List[str]:
    """Extrae todas las URLs únicas del texto."""
    patron = r'https?://[\w.-]+(?:[/\w.-]*)?'
    return list(set(re.findall(patron, texto)))

def procesar_texto(texto: str) -> Dict[str, List[str]]:
    """Procesa el texto y retorna un diccionario con los datos extraídos."""
    return {
        'emails': sorted(extraer_emails(texto)),
        'telefonos': sorted(extraer_telefonos(texto)),
        'urls': sorted(extraer_urls(texto)),
    }

def main() -> None:
    """Función principal: lee texto de ejemplo e imprime resultados."""
    texto_ejemplo = """
    Contacto principal: ana@uni.edu
    Para soporte técnico: soporte@empresa.com
    Visita nuestra web: https://www.ejemplo.com o https://otra-pagina.org
    Llámanos al +52 55 1234-5678 o al 55-8765-4321
    Email alternativo: info@ejemplo.com
    Sitio web: https://docs.python.org
    """
    
    resultados = procesar_texto(texto_ejemplo)
    
    print("=" * 50)
    print("RESULTADOS DEL SCRAPER")
    print("=" * 50)
    
    for tipo, valores in resultados.items():
        print(f"\n{tipo.upper()} ({len(valores)}):")
        for valor in valores:
            print(f"  - {valor}")

if __name__ == '__main__':
    main()
```
