## 1. Pilas (Stacks): LIFO

Una **pila** sigue el principio **LIFO** (Last In, First Out): el último elemento en entrar es el primero en salir. Como una pila de platos.

### Operaciones básicas

- **push**: agregar un elemento arriba
- **pop**: quitar el elemento de arriba
- **peek**: ver el elemento de arriba sin quitarlo
- **is_empty**: verificar si está vacía


```python
# Implementación de una pila usando una lista
class Pila:
    """Implementación de una pila (LIFO) usando una lista."""
    
    def __init__(self):
        self.items = []
    
    def push(self, item):
        """Agrega un elemento a la pila."""
        self.items.append(item)
    
    def pop(self):
        """Quita y retorna el elemento superior."""
        if self.is_empty():
            raise IndexError("pop from empty stack")
        return self.items.pop()
    
    def peek(self):
        """Retorna el elemento superior sin quitarlo."""
        if self.is_empty():
            raise IndexError("peek from empty stack")
        return self.items[-1]
    
    def is_empty(self):
        return len(self.items) == 0
    
    def size(self):
        return len(self.items)
    
    def __repr__(self):
        return f"Pila({self.items})"
```

```python
# Probar la pila
pila = Pila()
print(f"¿Vacía? {pila.is_empty()}")

pila.push("plato 1")
pila.push("plato 2")
pila.push("plato 3")
print(f"Pila: {pila}")
print(f"Tamaño: {pila.size()}")
print(f"Tope: {pila.peek()}")

print(f"\nPop: {pila.pop()}")
print(f"Pop: {pila.pop()}")
print(f"Pila: {pila}")
```

```python
# Caso de uso: verificar paréntesis balanceados
def verificar_parentesis(expresion):
    """Verifica que los paréntesis, llaves y corchetes estén balanceados."""
    pila = Pila()
    pares = {')': '(', ']': '[', '}': '{'}
    
    for caracter in expresion:
        if caracter in '([{':
            pila.push(caracter)
        elif caracter in ')]}':
            if pila.is_empty():
                return False
            if pila.pop() != pares[caracter]:
                return False
    return pila.is_empty()

# Probar
expresiones = ["(a + b)", "((a + b) * c)", "({[a + b]})", "(a + b", "a + b)", "((a)"]
for expr in expresiones:
    resultado = "✓ Balanceado" if verificar_parentesis(expr) else "✗ Desbalanceado"
    print(f"  '{expr}' → {resultado}")
```

## 2. Colas (Queues): FIFO

Una **cola** sigue el principio **FIFO** (First In, First Out): el primer elemento en entrar es el primero en salir. Como una cola de supermercado.

### Operaciones básicas

- **enqueue**: agregar un elemento al final
- **dequeue**: quitar el elemento del frente
- **front**: ver el elemento del frente
- **is_empty**: verificar si está vacía


```python
# Implementación de una cola usando collections.deque (eficiente)
from collections import deque

class Cola:
    """Implementación de una cola (FIFO) usando deque."""
    
    def __init__(self):
        self.items = deque()
    
    def enqueue(self, item):
        """Agrega un elemento al final de la cola."""
        self.items.append(item)
    
    def dequeue(self):
        """Quita y retorna el elemento del frente."""
        if self.is_empty():
            raise IndexError("dequeue from empty queue")
        return self.items.popleft()
    
    def front(self):
        """Retorna el elemento del frente sin quitarlo."""
        if self.is_empty():
            raise IndexError("front from empty queue")
        return self.items[0]
    
    def is_empty(self):
        return len(self.items) == 0
    
    def size(self):
        return len(self.items)
    
    def __repr__(self):
        return f"Cola({list(self.items)})")
```

```python
# Probar la cola
cola = Cola()
cola.enqueue("Ana")
cola.enqueue("Luis")
cola.enqueue("María")
print(f"Cola: {cola}")
print(f"Frente: {cola.front()}")

print(f"\nAtendiendo: {cola.dequeue()}")
print(f"Atendiendo: {cola.dequeue()}")
print(f"Cola: {cola}")
```

```python
# Caso de uso: simulación de atención al cliente
import time

class SistemaAtencion:
    """Simula un sistema de atención al cliente con cola."""
    
    def __init__(self):
        self.cola = Cola()
    
    def agregar_cliente(self, nombre):
        self.cola.enqueue(nombre)
        print(f"  → {nombre} se unió a la cola (posición {self.cola.size()})")
    
    def atender_siguiente(self):
        if self.cola.is_empty():
            print("  No hay clientes en espera.")
            return None
        cliente = self.cola.dequeue()
        print(f"  ← Atendiendo a {cliente}")
        return cliente

# Simulación
sistema = SistemaAtencion()
sistema.agregar_cliente("Ana")
sistema.agregar_cliente("Luis")
sistema.agregar_cliente("María")
print()
sistema.atender_siguiente()
sistema.atender_siguiente()
```

## 3. Colas de prioridad

Una **cola de prioridad** es similar a una cola, pero cada elemento tiene una **prioridad**. Los elementos con mayor prioridad se atienden primero.


```python
import heapq

class ColaPrioridad:
    """Cola de prioridad usando heapq (montículo binario)."""
    
    def __init__(self):
        self.items = []  # Heap de tuplas (prioridad, item)
    
    def enqueue(self, item, prioridad):
        """Agrega un elemento con su prioridad (menor = más prioritario)."""
        heapq.heappush(self.items, (prioridad, item))
    
    def dequeue(self):
        """Quita y retorna el elemento de mayor prioridad."""
        if self.is_empty():
            raise IndexError("dequeue from empty priority queue")
        return heapq.heappop(self.items)[1]
    
    def is_empty(self):
        return len(self.items) == 0
    
    def size(self):
        return len(self.items)
```

```python
# Probar cola de prioridad
cola = ColaPrioridad()
cola.enqueue("Tarea normal", prioridad=5)
cola.enqueue("Tarea urgente", prioridad=1)
cola.enqueue("Tarea importante", prioridad=3)

print("Procesando por prioridad:")
while not cola.is_empty():
    tarea = cola.dequeue()
    print(f"  → {tarea}")
```

## 4. Árboles y árboles binarios

Un **árbol** es una estructura jerárquica con nodos conectados. Cada nodo tiene un **padre** (excepto la raíz) e **hijos**.

### Árbol binario básico


```python
class Nodo:
    """Nodo de un árbol binario."""
    
    def __init__(self, valor):
        self.valor = valor
        self.izquierdo = None
        self.derecho = None
    
    def __repr__(self):
        return f"Nodo({self.valor})"

# Crear un árbol:
#         1
#        / \
#       2   3
#      / \   \
#     4   5   6
raiz = Nodo(1)
raiz.izquierdo = Nodo(2)
raiz.derecho = Nodo(3)
raiz.izquierdo.izquierdo = Nodo(4)
raiz.izquierdo.derecho = Nodo(5)
raiz.derecho.derecho = Nodo(6)

print(f"Raíz: {raiz.valor}")
print(f"Hijos: {raiz.izquierdo.valor}, {raiz.derecho.valor}")
```

```python
# Recorridos de un árbol binario
def preorden(nodo):
    """Raíz → Izquierdo → Derecho."""
    if nodo is None:
        return
    print(nodo.valor, end=" ")
    preorden(nodo.izquierdo)
    preorden(nodo.derecho)

def inorden(nodo):
    """Izquierdo → Raíz → Derecho."""
    if nodo is None:
        return
    inorden(nodo.izquierdo)
    print(nodo.valor, end=" ")
    inorden(nodo.derecho)

def postorden(nodo):
    """Izquierdo → Derecho → Raíz."""
    if nodo is None:
        return
    postorden(nodo.izquierdo)
    postorden(nodo.derecho)
    print(nodo.valor, end=" ")

print(f"Preorden: ", end="")
preorden(raiz)
print(f"\nInorden: ", end="")
inorden(raiz)
print(f"\nPostorden: ", end="")
postorden(raiz)
```

## 5. Árboles binarios de búsqueda (BST)

En un **BST**, para cada nodo:
- Todos los valores del subárbol izquierdo son menores.
- Todos los valores del subárbol derecho son mayores.

Esto permite búsquedas en **O(log n)** en promedio.


```python
class BST:
    """Árbol binario de búsqueda."""
    
    def __init__(self):
        self.raiz = None
    
    def insertar(self, valor):
        """Inserta un valor en el árbol."""
        self.raiz = self._insertar(self.raiz, valor)
    
    def _insertar(self, nodo, valor):
        if nodo is None:
            return Nodo(valor)
        if valor < nodo.valor:
            nodo.izquierdo = self._insertar(nodo.izquierdo, valor)
        else:
            nodo.derecho = self._insertar(nodo.derecho, valor)
        return nodo
    
    def buscar(self, valor):
        """Busca un valor en el árbol."""
        return self._buscar(self.raiz, valor)
    
    def _buscar(self, nodo, valor):
        if nodo is None or nodo.valor == valor:
            return nodo is not None
        if valor < nodo.valor:
            return self._buscar(nodo.izquierdo, valor)
        return self._buscar(nodo.derecho, valor)
```

```python
# Crear BST e insertar valores
bst = BST()
valores = [5, 3, 7, 1, 4, 6, 8]
for v in valores:
    bst.insertar(v)

# Buscar
print(f"¿Está 4? {bst.buscar(4)}")
print(f"¿Está 9? {bst.buscar(9)}")

# Inorden de un BST siempre da valores ordenados
print(f"\nInorden (ordenado):", end=" ")
inorden(bst.raiz)
```

## 6. Grafos y recorridos BFS/DFS

Un **grafo** es un conjunto de **nodos** (vértices) conectados por **aristas**. Puede ser dirigido o no dirigido.

### Representación con diccionario


```python
class Grafo:
    """Grafo simple con lista de adyacencia."""
    
    def __init__(self, dirigido=False):
        self.adj = {}
        self.dirigido = dirigido
    
    def agregar_arista(self, u, v):
        """Agrega una arista entre u y v."""
        if u not in self.adj:
            self.adj[u] = []
        if v not in self.adj:
            self.adj[v] = []
        self.adj[u].append(v)
        if not self.dirigido:
            self.adj[v].append(u)
    
    def vecinos(self, nodo):
        return self.adj.get(nodo, [])
    
    def __repr__(self):
        return f"Grafo({self.adj})"
```

```python
# Crear un grafo: mapa de ciudades
mapa = Grafo()
mapa.agregar_arista("CDMX", "Querétaro")
mapa.agregar_arista("CDMX", "Toluca")
mapa.agregar_arista("Querétaro", "San Luis Potosí")
mapa.agregar_arista("Querétaro", "Guadalajara")
mapa.agregar_arista("Toluca", "Puebla")
mapa.agregar_arista("Puebla", "Veracruz")

print(f"Mapa: {mapa}")
print(f"\nVecinos de CDMX: {mapa.vecinos('CDMX')}")
```

```python
# BFS (Búsqueda en Amplitud): usa una cola
def bfs(grafo, inicio):
    """Recorrido BFS desde un nodo inicial."""
    visitados = set()
    cola = deque([inicio])
    visitados.add(inicio)
    resultado = []
    
    while cola:
        nodo = cola.popleft()
        resultado.append(nodo)
        for vecino in grafo.vecinos(nodo):
            if vecino not in visitados:
                visitados.add(vecino)
                cola.append(vecino)
    return resultado

print(f"BFS desde CDMX: {bfs(mapa, 'CDMX')}")
```

```python
# DFS (Búsqueda en Profundidad): usa una pila (recursión)
def dfs(grafo, inicio, visitados=None):
    """Recorrido DFS desde un nodo inicial."""
    if visitados is None:
        visitados = set()
    visitados.add(inicio)
    resultado = [inicio]
    for vecino in grafo.vecinos(inicio):
        if vecino not in visitados:
            resultado.extend(dfs(grafo, vecino, visitados))
    return resultado

print(f"DFS desde CDMX: {dfs(mapa, 'CDMX')}")
```
