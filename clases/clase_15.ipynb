````python
# 1. "Beautiful is better than ugly"
# ❌ Feo
resultado = []
for i in range(10):
    resultado.append(i * 2)

# ✓ Hermoso (con comprensión de listas)
resultado = [i * 2 for i in range(10)]
print(resultado)
````

````python
# 2. "Explicit is better than implicit"
# ❌ Implícito (depende de estado global)
contador = 0
def incrementar():
    global contador
    contador += 1

# ✓ Explícito (pasa el estado como parámetro)
def incrementar(contador):
    return contador + 1

print(incrementar(0))
````

````python
# 3. "Simple is better than complex"
# ❌ Complejo: una función que hace todo
def procesar_datos(datos, opcion, formato, validar, limpiar):
    if validar and limpiar:
        datos = [d.strip() for d in datos if d]
    if opcion == 'A':
        return sorted(datos)
    elif opcion == 'B':
        return {d: len(d) for d in datos}
    return datos

# ✓ Simple: varias funciones pequeñas
def limpiar_datos(datos):
    return [d.strip() for d in datos if d]

def ordenar(datos):
    return sorted(datos)

def contar_caracteres(datos):
    return {d: len(d) for d in datos}

print(limpiar_datos(["  hola  ", " mundo", ""]))
````

````python
# 4. "Complex is better than complicated"
# A veces la complejidad es necesaria; lo que evitamos es la complicación
# ❌ Complicado: un truco "clever" difícil de entender
resultado = list(map(lambda x: x.upper(), filter(lambda x: x.startswith('A'), nombres)))

# ✓ Complejo pero claro: cada paso es explícito
def nombres_con_a(nombres):
    return [n for n in nombres if n.startswith('A')]

def en_mayusculas(nombres):
    return [n.upper() for n in nombres]

nombres = ["Ana", "Bob", "Alberto", "Carmen"]
print(en_mayusculas(nombres_con_a(nombres)))
````

````python
# 5. "Flat is better than nested"
# ❌ Anidado: if dentro de if dentro de if
def procesar(usuario, admin, activo):
    if usuario:
        if admin:
            if activo:
                return "Acceso completo"
            else:
                return "Acceso parcial"
        else:
            return "Sin acceso"
    return "No existe"

# ✓ Plano: guard clauses (retornos tempranos)
def procesar(usuario, admin, activo):
    if not usuario:
        return "No existe"
    if not admin:
        return "Sin acceso"
    if not activo:
        return "Acceso parcial"
    return "Acceso completo"

print(procesar("Ana", True, True))
````

````python
# 6. "Sparse is better than dense"
# ❌ Denso: todo en una línea
def es_valido(usuario): return usuario and usuario.activo and usuario.edad >= 18 and usuario.pais in ['MX', 'ES']

# ✓ Espaciado: lógica clara con variables
def es_valido(usuario):
    if not usuario:
        return False
    if not usuario.activo:
        return False
    if usuario.edad < 18:
        return False
    if usuario.pais not in ['MX', 'ES']:
        return False
    return True
````

````python
# 7. "Readability counts"
# ❌ Difícil de leer
x=lambda a,b,c:a+b*c-a/b if a else c

# ✓ Legible
def calcular(a, b, c):
    if a:
        return a + b * c - a / b
    return c

print(calcular(10, 2, 3))
````

````python
# 8. "Special cases aren't special enough to break the rules"
# ❌ Romper la convención "sin razón"
class MiClase:
    def __init__(self, x):
        self.x = x
    def metodo1(self): pass
    def MetodoRaro(self): pass  # Mayúscula para "ser especial"

# ✓ Consistente con PEP 8
class MiClase:
    def __init__(self, x):
        self.x = x
    def metodo_uno(self): pass
    def metodo_dos(self): pass
````

````python
# 9. "Although practicality beats purity"
# A veces hay que romper las reglas por pragmatismo
# Ejemplo: usar un nombre más corto para variables de uso único
# ❌ Demasiado purista
for indice_actual_del_bucle in range(10):
    print(f"Iteración {indice_actual_del_bucle}")

# ✓ Pragmático
for i in range(10):
    print(f"Iteración {i}")
````

````python
# 10. "Errors should never pass silently"
# ❌ Silenciar errores
try:
    resultado = operacion_peligrosa()
except:
    pass  # No hacer nada

# ✓ Manejar explícitamente
try:
    resultado = operacion_peligrosa()
except ValueError as e:
    print(f"Error específico: {e}")
    resultado = None
````

````python
# 11. "Unless explicitly silenced"
# A veces queremos silenciar errores a propósito, pero debe ser EXPLÍCITO
import contextlib

# Silenciar explícitamente con un context manager
@contextlib.contextmanager
def silenciar_errores():
    try:
        yield
    except Exception:
        pass

# Uso: cuando sabemos que el error es esperado
with silenciar_errores():
    # Esto fallará silenciosamente porque sabemos que no importa
    archivo = open("archivo_que_no_existe.txt", "r")

print("Continuamos sin problema")
````

````python
# 12. "In the face of ambiguity, refuse the temptation to guess"
# ❌ Adivinar
def obtener_valor(config):
    # ¿Qué pasa si config es None? ¿Si no tiene la clave?
    return config.valor or 0  # Asume 0 si es None o 0

# ✓ Ser explícito
def obtener_valor(config, default=0):
    if config is None:
        return default
    if not hasattr(config, 'valor'):
        return default
    return config.valor

class Config:
    pass

print(obtener_valor(None))
````

````python
# 17. "If the implementation is hard to explain, it's a bad idea"
# ❌ "Clever" pero difícil de explicar
resultado = list(map(lambda x: x[0], sorted(enumerate(datos), key=lambda x: x[1])))

# ✓ Claro y fácil de explicar
def obtener_primeros_elementos_ordenados(datos):
    """Retorna los primeros elementos de datos ordenados por su valor."""
    pares_con_indice = [(i, v) for i, v in enumerate(datos)]
    pares_ordenados = sorted(pares_con_indice, key=lambda par: par[1])
    return [indice for indice, _ in pares_ordenados]
````

````python
# 18. "If the implementation is easy to explain, it may be a good idea"
# ✓ Solución clara con buena documentación
def calcular_promedio_ponderado(valores, pesos):
    """Calcula el promedio ponderado de una lista de valores.
    
    Args:
        valores: Lista de valores numéricos.
        pesos: Lista de pesos (mismo tamaño que valores).
    
    Returns:
        El promedio ponderado, o None si las listas están vacías.
    """
    if not valores or not pesos:
        return None
    if len(valores) != len(pesos):
        raise ValueError("Las listas deben tener el mismo tamaño")
    
    suma_ponderada = sum(v * p for v, p in zip(valores, pesos))
    suma_pesos = sum(pesos)
    return suma_ponderada / suma_pesos

print(calcular_promedio_ponderado([9, 8, 10], [3, 2, 5]))
````

## 2. ¿Qué es un PEP?

Un **PEP** (Python Enhancement Proposal) es un documento que describe una propuesta para mejorar Python. Los PEPs son el mecanismo formal de evolución del lenguaje.

### El proceso

1. **Borrador:** cualquiera puede proponer una idea.
2. **Discusión:** la comunidad debate en listas de correo y GitHub.
3. **Aceptación:** un **PEP Delegado** (actualmente hay varios) acepta o rechaza.
4. **Implementación:** se implementa en una versión de Python.

### Tipos de PEPs

- **Standards Track:** nuevas funcionalidades del lenguaje (PEP 8, PEP 20, PEP 484).
- **Informational:** guías e información (PEP 7, PEP 8).
- **Process:** sobre el proceso de desarrollo mismo (PEP 1, PEP 12).


## 3. PEP 8 — Style Guide for Python Code

El **PEP 8** es la guía de estilo oficial de Python. No es obligatoria, pero seguirla hace que tu código sea legible para otros pythonistas.

### Reglas principales (resumen)

| Categoría | Regla | Ejemplo |
|---|---|---|
| **Indentación** | 4 espacios por nivel | `if x:\\n    print(x)` |
| **Longitud de línea** | Máximo 79 caracteres (estricto) o 99 (relajado) | Dividir líneas largas |
| **Nombres de variables** | `snake_case` | `mi_variable = 10` |
| **Nombres de clases** | `PascalCase` | `class MiClase:` |
| **Nombres de constantes** | `UPPER_SNAKE_CASE` | `PI = 3.14159` |
| **Funciones y variables privadas** | Prefijo `_` | `_mi_funcion()` |
| **Imports** | Uno por línea, agrupados | `import os\\nimport sys` |
| **Espacios** | Alrededor de operadores | `x = 1 + 2` (no `x=1+2`) |
| **Comentarios** | Completar oraciones | `# Calcular el promedio` |


````python
# ❌ Mal estilo (viola PEP 8)
def MiFuncion(x,y):
    PI=3.14159
    if x>0:return x*PI
    else:return 0

# ✓ Buen estilo (sigue PEP 8)
PI = 3.14159

def calcular_area(radio):
    """Calcula el área de un círculo dado su radio."""
    if radio > 0:
        return radio * PI
    return 0

print(calcular_area(5))
````

## 5. PEP 257 — Docstring Conventions

El **PEP 257** define convenciones para escribir docstrings (cadenas de documentación al inicio de módulos, clases y funciones).

### Estilos comunes

- **Google style:** usa secciones como `Args:`, `Returns:`, `Raises:`.
- **NumPy style:** usa bloques con guiones `---`.
- **Sphinx/reST:** usa campos como `:param x: description`.


````python
# Estilo Google (recomendado por su legibilidad)
def calcular_promedio_ponderado(valores, pesos):
    """Calcula el promedio ponderado de una lista de valores.
    
    Args:
        valores: Lista de valores numéricos.
        pesos: Lista de pesos (mismo tamaño que valores).
    
    Returns:
        float: El promedio ponderado.
    
    Raises:
        ValueError: Si las listas tienen tamaños diferentes.
    
    Examples:
        >>> calcular_promedio_ponderado([9, 8, 10], [3, 2, 5])
        9.1
    """
    if len(valores) != len(pesos):
        raise ValueError("Las listas deben tener el mismo tamaño")
    suma_ponderada = sum(v * p for v, p in zip(valores, pesos))
    return suma_ponderada / sum(pesos)

# Estilo NumPy
def dividir_seguro(a, b):
    """Divide dos números de forma segura.
    
    Parameters
    ----------
    a : float
        El numerador.
    b : float
        El denominador.
    
    Returns
    -------
    float
        El resultado de la división, o infinito si b es cero.
    """
    if b == 0:
        return float('inf')
    return a / b

print(calcular_promedio_ponderado([9, 8, 10], [3, 2, 5]))
````

## 7. Cómo se lee un PEP

Los PEPs son documentos técnicos formales. Aquí va una guía rápida:

1. **Abstract (Resumen):** qué propone el PEP en un párrafo.
2. **Motivation (Motivación):** por qué es necesario.
3. **Specification (Especificación):** detalles técnicos exactos.
4. **Backwards Compatibility:** si rompe código existente.
5. **Reference Implementation:** enlace al código de prueba.
6. **Rejected Ideas:** ideas que se consideraron y descartaron.

### Ejemplo: leer PEP 8 rápidamente

Si solo quieres las reglas esenciales, lee:
- "Code Lay-out" (indentación, líneas, imports).
- "Naming Conventions" (cómo nombrar variables, clases, funciones).
- "Comments" y "Documentation Strings".


## 8. Ejercicio Práctico: Refactorizar aplicando el Zen

Vas a recibir un código que viola varios principios del Zen. Tu trabajo es refactorizarlo para que siga las buenas prácticas.

**Instrucciones:**

Crea un archivo `refactorizar.py` con una función `procesar_calificaciones(notas)` que:
1. Reciba una lista de tuplas `(nombre, notas)`.
2. Calcule el promedio de cada estudiante.
3. Retorne una lista de tuplas `(nombre, promedio, estado)` donde `estado` es `"Aprobado"` si promedio >= 7, `"Reprobado"` en caso contrario.
4. Ordene la lista de mayor a menor promedio.

**Pistas:**
- Usa nombres descriptivos (Zen: "Readability counts").
- Divide en funciones pequeñas si es necesario (Zen: "Simple is better than complex").
- Usa guard clauses en vez de `if/else` anidados (Zen: "Flat is better than nested").
- Maneja la lista vacía explícitamente (Zen: "Errors should never pass silently").
- Documenta con docstring estilo Google (PEP 257).
- Sigue PEP 8: snake_case, 4 espacios, líneas < 79 caracteres.

**Restricción del ejercicio:**
- No uses librerías externas.
- Sigue el Zen de Python y PEP 8.


````python
"""
Solución al Ejercicio Práctico
Este código debe ir en un archivo refactorizar.py
"""

# ❌ Código inicial (viola el Zen y PEP 8)
def f(x):
    r=[]
    for i in x:
        avg=0
        for n in i[1]:avg=avg+n
        avg=avg/len(i[1])
        if avg>=7:s='Aprobado'
        else:s='Reprobado'
        r.append((i[0],avg,s))
    # Sin ordenar
    return r

# Versión corregida
def calcular_promedios(estudiantes):
    """Calcula el promedio de cada estudiante.
    
    Args:
        estudiantes: Lista de tuplas (nombre, lista_de_notas).
    
    Returns:
        Lista de tuplas (nombre, promedio).
    """
    if not estudiantes:
        return []
    return [
        (nombre, sum(notas) / len(notas))
        for nombre, notas in estudiantes
        if notas  # Ignorar estudiantes sin notas
    ]


def clasificar(promedio, minimo=7.0):
    """Clasifica un promedio como Aprobado o Reprobado."""
    return "Aprobado" if promedio >= minimo else "Reprobado"


def procesar_calificaciones(estudiantes, minimo=7.0):
    """Procesa calificaciones: calcula promedios y clasifica.
    
    Args:
        estudiantes: Lista de tuplas (nombre, lista_de_notas).
        minimo: Calificación mínima para aprobar (default 7.0).
    
    Returns:
        Lista de tuplas (nombre, promedio, estado) ordenada
        de mayor a menor promedio.
    
    Examples:
        >>> procesar_calificaciones([("Ana", [9, 10]), ("Luis", [5, 6])])
        [('Ana', 9.5, 'Aprobado'), ('Luis', 5.5, 'Reprobado')]
    """
    # Guard clause: manejar lista vacía
    if not estudiantes:
        return []
    
    # Calcular promedios usando la función auxiliar
    promedios = calcular_promedios(estudiantes)
    
    # Clasificar y ordenar
    resultados = [
        (nombre, promedio, clasificar(promedio, minimo))
        for nombre, promedio in promedios
    ]
    return sorted(resultados, key=lambda x: x[1], reverse=True)
````

````python
# Probar
datos = [
    ("Ana", [9.0, 10.0, 8.5]),
    ("Luis", [6.0, 5.5, 7.0]),
    ("María", [10.0, 9.5, 9.0]),
    ("Pedro", []),  # Sin notas
]
resultado = procesar_calificaciones(datos)
print("--- Calificaciones procesadas ---")
for nombre, promedio, estado in resultado:
    print(f"  {nombre}: {promedio:.2f} - {estado}")

print(f"\n--- Lista vacía ---")
print(procesar_calificaciones([]))
````
